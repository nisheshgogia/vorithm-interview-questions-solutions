# Category 1 — Basic SELECT, Filtering & Sorting

## 1. Employees Whose Name Starts With 'A'

**Difficulty:** 🟢 Easy

**Subtopic:** LIKE / Filtering

---

**Table: employee_salary**
```
+------------------+---------+
| Column Name      | Type    |
+------------------+---------+
| did              | int     |
| name_of_employee | varchar |
| salary           | int     |
+------------------+---------+
```

- `did` represents the department ID.
- `name_of_employee` represents the employee's name.
- `salary` represents the employee's salary.

Each department may contain multiple employees.

---

**Problem**

Write a SQL query to return the name of the employee whose name starts with the letter **'A'**.

Return the result in any order.

**Example 1**

**Input**

employee_salary table:
```
+-----+------------------+--------+
| did | name_of_employee | salary |
+-----+------------------+--------+
| 1   | Alice            | 50000  |
| 1   | Bob              | 65000  |
| 1   | Carol            | 60000  |
| 2   | David            | 70000  |
| 2   | Emma             | 72000  |
| 2   | Frank            | 68000  |
| 3   | Anna             | 55000  |
| 3   | George           | 52000  |
| 3   | Henry            | 51000  |
+-----+------------------+--------+
```

**Output**
```
+------------------+
| name_of_employee |
+------------------+
| Alice            |
| Anna             |
+------------------+
```

---



````python
# 👀 Preview the table
print("employee_salary table:")
pd.read_sql_query("SELECT * FROM employee_salary", conn)
````

SOLUTION

````python

query = """
SELECT name_of_employee
FROM employee_salary
WHERE name_of_employee LIKE 'A%'
"""

pd.read_sql_query(query, conn)
````

**Explanation**

- `WHERE` filters rows before any grouping or aggregation.
- `LIKE 'A%'` matches any name that **starts with the letter 'A'** — the `%` is a wildcard that means *"anything can follow"*.
- Only `Alice` and `Anna` satisfy this condition from the table.

# Q2. Employees Hired in the Last 3 Months

## 2. Employees Hired in the Last 3 Months

**Difficulty:** 🟢 Easy

**Subtopic:** DATE Filtering

---

**Table: employees**
```
+------------+---------+
| Column Name| Type    |
+------------+---------+
| id         | int     |
| name       | varchar |
| salary     | int     |
| dep_id     | int     |
| manager_id | int     |
| hire_date  | date    |
+------------+---------+
```

- `id` represents the employee ID.
- `name` represents the employee's name.
- `hire_date` represents the date the employee joined.

---

**Problem**

Write a SQL query to find all employees who were hired in the **last 3 months** from today.

Return all columns for matching employees.

**Example 1**

**Input**

employees table:
```
+----+-------+--------+--------+------------+------------+
| id | name  | salary | dep_id | manager_id | hire_date  |
+----+-------+--------+--------+------------+------------+
| 1  | John  | 60000  | 1      | NULL       | 2023-01-01 |
| 2  | Alice | 80000  | 1      | 1          | 2023-03-01 |
| 3  | Bob   | 75000  | 2      | 1          | 2023-05-01 |
| 4  | David | 50000  | 2      | 3          | 2024-01-01 |
+----+-------+--------+--------+------------+------------+
```

**Output** *(assuming today is 2024-03-01)*
```
+----+-------+--------+--------+------------+------------+
| id | name  | salary | dep_id | manager_id | hire_date  |
+----+-------+--------+--------+------------+------------+
| 4  | David | 50000  | 2      | 3          | 2024-01-01 |
+----+-------+--------+--------+------------+------------+
```

(View Table):

````python
# 👀 Preview the table
print("employees table:")
pd.read_sql_query("SELECT * FROM employees", conn)
````

CODE (Solution):

````python
# ✅ Solution
query = """
SELECT *
FROM employees
WHERE hire_date >= DATE('now', '-3 months')
"""

pd.read_sql_query(query, conn)
````

> **Note:** Since our sample data uses older dates,

> we demonstrate the concept by replacing `DATE('now', '-3 months')`

> with a fixed reference date. In a real scenario `DATE('now')`

> would always refer to today's date automatically.

# Q3. Display Total Salary After 10% Increment

## 3. Display Total Salary After 10% Increment

**Difficulty:** 🟢 Easy

**Subtopic:** Arithmetic / Derived Columns

---

**Table: employee_salary**
```
+------------------+---------+
| Column Name      | Type    |
+------------------+---------+
| did              | int     |
| name_of_employee | varchar |
| salary           | int     |
+------------------+---------+
```

- `did` represents the department ID.
- `name_of_employee` represents the employee's name.
- `salary` represents the employee's salary.

---

**Problem**

Write a SQL query to display the **name** and **new salary** of each employee after adding a **10% increment** to their existing salary.

**Example 1**

**Input**

employee_salary table:
```
+-----+------------------+--------+
| did | name_of_employee | salary |
+-----+------------------+--------+
| 1   | Alice            | 50000  |
| 1   | Bob              | 65000  |
| 1   | Carol            | 60000  |
| 2   | David            | 70000  |
| 2   | Emma             | 72000  |
| 2   | Frank            | 68000  |
| 3   | Anna             | 55000  |
| 3   | George           | 52000  |
| 3   | Henry            | 51000  |
+-----+------------------+--------+
```

**Output**
```
+------------------+------------+
| name_of_employee | new_salary |
+------------------+------------+
| Alice            | 55000.0    |
| Bob              | 71500.0    |
| Carol            | 66000.0    |
| David            | 77000.0    |
| Emma             | 79200.0    |
| Frank            | 74800.0    |
| Anna             | 60500.0    |
| George           | 57200.0    |
| Henry            | 56100.0    |
+------------------+------------+
```

CODE (View Table):

````python
# 👀 Preview the table
print("employee_salary table:")
pd.read_sql_query("SELECT * FROM employee_salary", conn)
````

CODE (Solution):

````python
# ✅ Solution
query = """
SELECT name_of_employee,
       salary * 1.10 AS new_salary
FROM employee_salary
"""

pd.read_sql_query(query, conn)
````

**Explanation**

- `salary * 1.10` multiplies each employee's salary by 1.10 which adds a 10% increment.
- `AS new_salary` renames the calculated column to `new_salary` in the output.
- No `WHERE` or `GROUP BY` is needed — this applies to every row in the table.

(View Tables):

````python
# 👀 Preview Table 1
print("employees table:")
pd.read_sql_query("SELECT * FROM employees", conn)
````

````python
# 👀 Preview Table 2
print("department table:")
pd.read_sql_query("SELECT * FROM department", conn)
````

CODE (Solution):

````python
# ✅ Solution
query = """
SELECT e.*
FROM employees e
JOIN department d ON e.dep_id = d.id
WHERE d.name NOT IN ('HR', 'Admin')
"""

pd.read_sql_query(query, conn)
````

**Explanation**

- `JOIN department d ON e.dep_id = d.id` connects employees to their department name.
- `WHERE d.name NOT IN ('HR', 'Admin')` excludes any employee whose department is HR or Admin.
- `NOT IN` is the cleanest way to exclude multiple values at once.

**Example 1**

**Input**

employees table:
```
+----+-------+
| id | name  |
+----+-------+
```

**Explanation**

- `HAVING COUNT(*) > 1` keeps only those names that appear **more than once** — i.e., duplicates.
- `HAVING` is used instead of `WHERE` because we are filtering on an aggregated value.

## 7. Companies With At Least 2 Users Speaking English AND German

**Difficulty:** 🟡 Medium

**Subtopic:** GROUP BY + HAVING + DISTINCT

---

**Table: users**
```
+-------------+---------+
| Column Name | Type    |
+-------------+---------+
| user_id     | int     |
| company     | varchar |
| language    | varchar |
+-------------+---------+
```

- `user_id` represents the unique user.
- `company` represents the company the user belongs to.
- `language` represents the language spoken by the user.

---

**Problem**

Write a SQL query to return the names of companies that have **at least one English speaker AND at least one German speaker**.

**Example 1**

**Input**

users table:
```
+---------+---------+----------+
| user_id | company | language |
+---------+---------+----------+
| 1       | Google  | English  |
| 2       | Google  | German   |
| 3       | Google  | French   |
| 4       | Meta    | English  |
| 5       | Meta    | Spanish  |
| 6       | Apple   | English  |
| 7       | Apple   | German   |
+---------+---------+----------+
```

**Output**
```
+---------+
| company |
+---------+
| Google  |
| Apple   |
+---------+
```

CODE (Solution):

````python
# ✅ Solution
query = """
SELECT company
FROM users
WHERE language IN ('English', 'German')
GROUP BY company
HAVING COUNT(DISTINCT language) = 2
"""

pd.read_sql_query(query, conn)
````

**Explanation**

- `WHERE language IN ('English', 'German')` filters only English and German speakers first.
- `GROUP BY company` groups the filtered rows by company.
- `COUNT(DISTINCT language)` counts how many **distinct** languages (English, German) each company has.
- `HAVING COUNT(DISTINCT language) = 2` ensures the company has **both** English and German speakers — not just one of them.
