# 1. Maximum Salary by Department

**Difficulty:** 🟢 Easy  
**Subtopic:** GROUP BY

---

## Table: employee_salary

```
+------------------+---------+
| Column Name      | Type    |
+------------------+---------+
| did              | int     |
| name_of_employee | varchar |
| salary           | int     |
+------------------+---------+
```

- `did` represents the department ID.  
- `name_of_employee` represents the employee's name.  
- `salary` represents the employee's salary.  

Each department may contain multiple employees.

---



## Problem

Write a SQL query to **return the maximum salary in each department**.

Return the result table containing the **department ID and the maximum salary** in that department.

Return the result table in **any order**.

---

## Example 1

### Input

employee_salary table:

```
+-----+------------------+--------+
| did | name_of_employee | salary |
+-----+------------------+--------+
| 1   | Alice            | 50000  |
| 1   | Bob              | 65000  |
| 1   | Carol            | 60000  |
| 2   | David            | 70000  |
| 2   | Emma             | 72000  |
| 3   | Frank            | 55000  |
+-----+------------------+--------+
```

---

### Output

```
+-----+------------+
| did | max_salary |
+-----+------------+
| 1   | 65000      |
| 2   | 72000      |
| 3   | 55000      |
+-----+------------+
```

---



---



Preview the table

````python
# 👀 Preview the table
print("employee_salary table:")
pd.read_sql_query("SELECT * FROM employee_salary", conn)
````

SOLUTION

````python
query = """
SELECT did,
       MAX(salary) AS max_salary
FROM employee_salary
GROUP BY did
"""

pd.read_sql_query(query,conn)
````

# Explanation

- `GROUP BY did` groups employees by department.
- `MAX(salary)` finds the highest salary within each department.

GROUP BY did groups employees by department.

AVG(salary) calculates the average salary within each department.

# 5. Departments with More Than One Employee

**Difficulty:** 🟢 Easy  
**Subtopic:** GROUP BY / HAVING

```
## Table: employee_salary

+------------------+---------+
| Column Name      | Type    |
+------------------+---------+
| did              | int     |
| name_of_employee | varchar |
| salary           | int     |
+------------------+---------+

- `did` represents the department ID.  
- `name_of_employee` represents the employee's name.  
- `salary` represents the employee's salary.  

Each department may contain multiple employees.

```
## Problem

Write a SQL query to **find departments that have more than one employee**.
Return the result table containing the **department ID and the number of employees** in that department.
Return the result table in **any order**.

```

Example 1
Input
employee_salary table:

+-----+------------------+--------+
| did | name_of_employee | salary |
+-----+------------------+--------+
| 1   | Alice            | 50000  |
| 1   | Bob              | 65000  |
| 1   | Carol            | 60000  |
| 2   | David            | 70000  |
| 2   | Emma             | 72000  |
| 3   | Frank            | 55000  |
+-----+------------------+--------+



Output
+-----+----------------+
| did | employee_count |
+-----+----------------+
| 1   | 3              |
| 2   | 2              |
+-----+----------------+

```



Preview the table

````python
# 👀 Preview — used for all questions in this category
print("employee_salary table:")
pd.read_sql_query("SELECT * FROM employee_salary", conn)
````

Solution

````python
query = """
SELECT
    did,
    COUNT(name_of_employee) AS employee_count
FROM employee_salary
GROUP BY did
HAVING COUNT(name_of_employee) > 1
"""

pd.read_sql_query(query, conn)
````

Problem

Return the result table containing:
department ID
average salary of that department
Return only the department with the lowest average salary.

Example 1
Input
```
employee_salary table:
```
+-----+------------------+--------+
| did | name_of_employee | salary |
+-----+------------------+--------+
| 1   | Alice            | 50000  |
| 1   | Bob              | 65000  |
| 1   | Carol            | 60000  |
| 2   | David            | 70000  |
| 2   | Emma             | 72000  |

Solution

````python
query = """
SELECT
    did,
    AVG(salary) AS avg_salary
FROM employee_salary
GROUP BY did
ORDER BY avg_salary ASC
LIMIT 1
"""

pd.read_sql_query(query, conn)
````
