# 1. Convert Categorical Labels to Numbers (Label Encoding)
# Difficulty: 🟢 LowProblem Statement:

Given a list of strings labels, return a list of integers where each unique label is mapped to a unique integer starting from 0 in the order of first appearance (label encoding).This is a fundamental preprocessing step in machine learning where categorical variables need to be converted to numerical form.
```
Example 1:
Input: labels = ["cat", "dog", "cat", "bird"]
Output: [0, 1, 0, 2]
Explanation:
"cat" → 0 (first appearance)
"dog" → 1 (second unique label)
"cat" → 0 (already mapped)
"bird" → 2 (third unique label)
```
```
Example 2:
Input: labels = ["red", "blue", "red", "green", "blue"]
Output: [0, 1, 0, 2, 1]
Explanation:
"red" → 0
"blue" → 1
"red" → 0 (already seen)
"green" → 2
"blue" → 1 (already seen)Example 3:
Input: labels = []
Output: []Example 4:
Input: labels = ["a", "a", "a"]
Output: [0, 0, 0]

````python
#Brute Force Approach (Check Unique List for Each Element)
def label_encode_brute(labels):
    unique_labels = []
    result = []

    for label in labels:
        # Check if label already in unique_labels
        if label not in unique_labels:
            unique_labels.append(label)

        # Find index in unique_labels
        encoded = unique_labels.index(label)
        result.append(encoded)

    return result

print(label_encode_brute(["cat", "dog", "cat", "bird"]))
# Output: [0, 1, 0, 2]

print(label_encode_brute(["red", "blue", "red", "green", "blue"]))
# Output: [0, 1, 0, 2, 1]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : all labels same → "in" check always O(1) (first position)
#                  index() also O(1)
# avg   → O(n*k) : where k = average number of unique labels
#                  for each of n labels:
#                  - "in" check: O(k) on average (scan unique_labels)
#                  - index(): O(k) on average
#                  total ≈ n × 2k = O(n*k)
# worst → O(n^2) : all labels unique → k grows to n
#                  label n requires checking n unique labels
#                  total = 1 + 2 + 3 + ... + n = n(n+1)/2 = O(n^2)
# space → O(k)   : unique_labels stores k unique labels
#                  result stores n elements (output, not counted)

#best time complexity    - O(n)
#average time complexity - O(n*k) where k = unique labels
#worst time complexity   - O(n^2)
#space complexity        - O(k)
````

````python
# Optimized Approach (HashMap/Dictionary)
def label_encode_optimized(labels):
    label_to_int = {}
    result = []
    next_id = 0

    for label in labels:
        if label not in label_to_int:
            label_to_int[label] = next_id
            next_id += 1

        result.append(label_to_int[label])

    return result


labels = ["cat", "dog", "cat", "bird"]

print(label_encode_optimized(labels))
````

````python
# Alternative: More concise using get()
def label_encode_concise(labels):
    label_map = {}
    result = []

    for label in labels:
        if label not in label_map:
            label_map[label] = len(label_map)
        result.append(label_map[label])

    return result


labels = ["cat", "dog", "cat", "bird"]

print(label_encode_optimized(labels))
print(label_encode_concise(labels))
````

````python
# One-liner version (less readable but Pythonic)
def label_encode_oneliner(labels):
    mapping = {}
    return [mapping.setdefault(label, len(mapping)) for label in labels]

print(label_encode_optimized(["cat", "dog", "cat", "bird"]))
# Output: [0, 1, 0, 2]

print(label_encode_concise(["red", "blue", "red", "green", "blue"]))
# Output: [0, 1, 0, 2, 1]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : even all same labels, still process each once
#                  dictionary lookup/insert is O(1) average
# avg   → O(n)   : for each of n labels:
#                  - dictionary "in" check: O(1) average
#                  - dictionary insert/lookup: O(1) average
#                  total = n × O(1) = O(n)
# worst → O(n)   : all unique labels still O(n)
#                  hash collisions rare, still amortized O(1)
# space → O(k)   : label_to_int dictionary stores k unique labels

# KEY INSIGHT: Dictionary provides O(1) lookup/insert
# vs list's O(k) for "in" check and index()

# VISUALIZATION for ["cat", "dog", "cat", "bird"]:
# i=0: label="cat", not in map, map={"cat":0}, result=[0]
# i=1: label="dog", not in map, map={"cat":0,"dog":1}, result=[0,1]
# i=2: label="cat", in map, result=[0,1,0]
# i=3: label="bird", not in map, map={"cat":0,"dog":1,"bird":2}, result=[0,1,0,2]

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(k) where k = unique labels


````

## 2. Compute Mean Squared Error (MSE)
**Difficulty: 🟢 Low**

**Problem Statement:**
Given two lists of floats y_true (actual values) and y_pred (predicted values) of the same length, return the mean squared error.

MSE is a common regression metric calculated as: MSE = (1/n) × Σ(y_true[i] - y_pred[i])²

**Example 1:**
```
Input: y_true = [3, -0.5, 2, 7], y_pred = [2.5, 0.0, 2, 8]
Output: 0.375
Explanation:
Errors: [0.5, -0.5, 0, -1]
Squared: [0.25, 0.25, 0, 1]
Mean: (0.25 + 0.25 + 0 + 1) / 4 = 1.5 / 4 = 0.375
```

**Example 2:**
```
Input: y_true = [1, 2, 3], y_pred = [1, 2, 3]
Output: 0.0
Explanation: Perfect predictions, all errors are 0
```

**Example 3:**
```
Input: y_true = [10], y_pred = [12]
Output: 4.0
Explanation: (10-12)² / 1 = 4

````python
#Brute Force Approach (Explicit Loop with Accumulator)
def mse_brute(y_true, y_pred):
    if len(y_true) == 0:
        return 0.0

    n = len(y_true)
    total_squared_error = 0.0

    for i in range(n):
        error = y_true[i] - y_pred[i]
        squared_error = error ** 2
        total_squared_error += squared_error

    mse = total_squared_error / n
    return mse

print(mse_brute([3, -0.5, 2, 7], [2.5, 0.0, 2, 8]))
# Output: 0.375

print(mse_brute([1, 2, 3], [1, 2, 3]))
# Output: 0.0

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : even if all errors zero, must compute all n differences
# avg   → O(n)   : loop runs n times, each iteration:
#                  - subtraction: O(1)
#                  - squaring: O(1)
#                  - addition: O(1)
#                  total = n × O(1) = O(n)
# worst → O(n)   : same operations regardless of values
# space → O(1)   : only total_squared_error, error, squared_error variables

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

````python
#Optimized Approach (List Comprehension + Built-in sum())
def mse_optimized(y_true, y_pred):
    if len(y_true) == 0:
        return 0.0

    n = len(y_true)
    # Compute all squared errors in one expression
    squared_errors = [(true - pred) ** 2
                      for true, pred in zip(y_true, y_pred)]

    return sum(squared_errors) / n

# More concise one-liner
def mse_oneliner(y_true, y_pred):
    n = len(y_true)
    return sum((true - pred) ** 2 for true, pred in zip(y_true, y_pred)) / n if n > 0 else 0.0

# NumPy-style (if allowed)
def mse_numpy_style(y_true, y_pred):
    """
    If numpy were allowed:
    import numpy as np
    return np.mean((np.array(y_true) - np.array(y_pred)) ** 2)
    """
    n = len(y_true)
    return sum((y_true[i] - y_pred[i]) ** 2 for i in range(n)) / n if n > 0 else 0.0

print(mse_optimized([3, -0.5, 2, 7], [2.5, 0.0, 2, 8]))
# Output: 0.375

print(mse_oneliner([1, 2, 3], [1, 2, 3]))
# Output: 0.0

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : must process all n elements regardless
# avg   → O(n)   : zip() creates iterator: O(n) to consume
#                  list comprehension: n iterations × O(1) = O(n)
#                  sum(): O(n) to sum n values
#                  total = O(n) + O(n) = O(n)
# worst → O(n)   : same regardless of input values
# space → O(n)   : squared_errors list stores n values
#                  (generator version would be O(1) space)

# KEY INSIGHT: No algorithmic improvement over brute force,
# but cleaner, more Pythonic code
# For true O(1) space, use generator: (... for ...)

# STEP-BY-STEP for [3, -0.5, 2, 7], [2.5, 0.0, 2, 8]:
# zip pairs: (3, 2.5), (-0.5, 0.0), (2, 2), (7, 8)
# differences: 0.5, -0.5, 0, -1
# squared: 0.25, 0.25, 0, 1
# sum: 1.5
# mean: 1.5 / 4 = 0.375 ✓

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n) for list, O(1) for generator


````

## 3. Compute Accuracy
**Difficulty: 🟢 Low**

**Problem Statement:**
Given two lists y_true (actual labels) and y_pred (predicted labels) of the same length, return the accuracy as a float between 0 and 1.

Accuracy = (Number of correct predictions) / (Total predictions)

**Example 1:**
```
Input: y_true = [0, 1, 1, 0], y_pred = [0, 1, 0, 0]
Output: 0.75
Explanation:
Matches: ✓ ✓ ✗ ✓
3 out of 4 correct → 3/4 = 0.75
```

**Example 2:**
```
Input: y_true = [1, 0, 1, 1, 0], y_pred = [1, 0, 0, 1, 0]
Output: 0.8
Explanation:
Matches: ✓ ✓ ✗ ✓ ✓
4 out of 5 correct → 4/5 = 0.8
```

**Example 3:**
```
Input: y_true = [1, 2, 3], y_pred = [1, 2, 3]
Output: 1.0
Explanation: Perfect predictions
```

**Example 4:**
```
Input: y_true = [0, 0], y_pred = [1, 1]
Output: 0.0
Explanation: All predictions wrong

````python
#Brute Force Approach (Explicit Loop with Counter)
def accuracy_brute(y_true, y_pred):
    if len(y_true) == 0:
        return 0.0

    n = len(y_true)
    correct = 0

    for i in range(n):
        if y_true[i] == y_pred[i]:
            correct += 1

    accuracy = correct / n
    return accuracy

print(accuracy_brute([0, 1, 1, 0], [0, 1, 0, 0]))
# Output: 0.75

print(accuracy_brute([1, 0, 1, 1, 0], [1, 0, 0, 1, 0]))
# Output: 0.8

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : even if all match, must check all n positions
# avg   → O(n)   : loop runs n times, each iteration:
#                  - comparison: O(1)
#                  - increment: O(1)
#                  total = n × O(1) = O(n)
# worst → O(n)   : same operations regardless of match/mismatch
# space → O(1)   : only correct counter variable

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

````python
# Optimized Approach (sum() with Generator Expression)
def accuracy_optimized(y_true, y_pred):
    if len(y_true) == 0:
        return 0.0

    n = len(y_true)
    correct = sum(true == pred for true, pred in zip(y_true, y_pred))

    return correct / n


y_true = [1, 0, 1, 1, 0]
y_pred = [1, 0, 0, 1, 0]

print(accuracy_optimized(y_true, y_pred))
````

````python
# Alternative: Using explicit list
def accuracy_list(y_true, y_pred):
    if len(y_true) == 0:
        return 0.0

    n = len(y_true)
    matches = [1 if y_true[i] == y_pred[i] else 0 for i in range(n)]
    return sum(matches) / n


y_true = [1, 0, 1, 1, 0]
y_pred = [1, 0, 0, 1, 0]

print(accuracy_list(y_true, y_pred))
````

````python
# One-liner
def accuracy_oneliner(y_true, y_pred):
    n = len(y_true)
    return sum(t == p for t, p in zip(y_true, y_pred)) / n if n > 0 else 0.0

print(accuracy_optimized([0, 1, 1, 0], [0, 1, 0, 0]))
# Output: 0.75

print(accuracy_oneliner([1, 0, 1, 1, 0], [1, 0, 0, 1, 0]))
# Output: 0.8

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : must check all n elements for matches
# avg   → O(n)   : zip() creates iterator: O(1) setup
#                  generator expression: n comparisons × O(1) = O(n)
#                  sum(): O(n) to sum n boolean values
#                  total = O(n)
# worst → O(n)   : same operations regardless of accuracy value
# space → O(1)   : generator expression doesn't store intermediate list
#                  (list version would be O(n))

# KEY INSIGHT: Python treats True as 1, False as 0
# sum(True, False, True) = sum(1, 0, 1) = 2

# STEP-BY-STEP for [0,1,1,0], [0,1,0,0]:
# zip pairs: (0,0), (1,1), (1,0), (0,0)
# comparisons: True, True, False, True
# as integers: 1, 1, 0, 1
# sum: 3
# accuracy: 3/4 = 0.75 ✓

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1) for generator, O(n) for list
````
