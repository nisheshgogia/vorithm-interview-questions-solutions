#1 . A dataset contains salary information stored as text such as ""$90K-$120K"". Extract the minimum salary, maximum salary, and average salary from the text value.

 (Difficulty: 🟢 Low)

```
Example:
Input ""$90K-$120K"".
Output: min_salary = 90000,
max_salary = 120000,
avg_salary = 105000.

````
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/nisheshgogia/vorithm-interview-questions-solutions/main/Python/data_analytics/dataanalyst.csv")

# Cleaning the Salary Estimate column
# Remove the extra text "(Glassdoor est.)"
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Glassdoor est.)', '', regex=False).str.strip()

# Split the salary range into two parts
df[['min_part', 'max_part']] = df['Salary Estimate'].str.split('-', expand=True)

# Remove symbols like $ and K
df['min_part'] = df['min_part'].str.replace('$', '', regex=False).str.replace('K', '', regex=False).str.strip()
df['max_part'] = df['max_part'].str.replace('$', '', regex=False).str.replace('K', '', regex=False).str.strip()

# Convert values to numeric
df['min_salary'] = pd.to_numeric(df['min_part'], errors='coerce')
df['max_salary'] = pd.to_numeric(df['max_part'], errors='coerce')

# Since salaries were in thousands (K), multiply by 1000
df['min_salary'] = df['min_salary'] * 1000
df['max_salary'] = df['max_salary'] * 1000

# Create average salary column
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2

# Remove temporary columns used for cleaning
df.drop(columns=['min_part', 'max_part'], inplace=True)

# Check the result
print(df[['Salary Estimate', 'min_salary', 'max_salary', 'avg_salary']].head(10))
````

# 3. Convert Salary to Numeric
# **(Difficulty: 🟢 Low)**

```
Problem:
Salary values are stored as strings such as "$85K", "$120K", or "$60K". Convert these values into numeric format so they can be used for analysis.

Example:
Input "$85K".
Output: 85000.

````
import pandas as pd

# Create a small example dataset
df = pd.DataFrame({
    'Salary': ['$85K', '$120K', '$60K', '$95K', '$110K']
})

# Convert salary values to numbers
numeric_salaries = []

for salary in df['Salary']:

    # Remove the dollar sign
    salary = salary.replace('$', '')

    # Remove the 'K'
    salary = salary.replace('K', '')

    # Convert the remaining value to float
    salary_value = float(salary)

    # Multiply by 1000 since K means thousand
    salary_value = salary_value * 1000

    # Store the result
    numeric_salaries.append(salary_value)

# Add the cleaned numeric salary column
df['Salary_Numeric'] = numeric_salaries

# Show the final dataframe
print(df)
````

# 5. Standardize Company Size
# (Difficulty: 🟢 Low)

```
Problem:
A dataset contains company size information written in text such as "51 to 200 employees" or "10000+ employees". Convert these values into standardized size categories.

Example:
Input: "51 to 200 employees"
Output: "Small"

````
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/nisheshgogia/vorithm-interview-questions-solutions/main/Python/data_analytics/dataanalyst.csv")

# Function to convert company size text
# into standardized size categories
def categorize_size(size_text):

    # If the value is missing, return 'Unknown'
    if pd.isna(size_text):
        return 'Unknown'

    # Convert text to lowercase to make comparison easier
    size_lower = str(size_text).lower()

    # Check different ranges and assign a category
    if '1 to 50' in size_lower:
        return 'Micro'

    elif '51 to 200' in size_lower:
        return 'Small'

    elif '201 to 500' in size_lower or '201 to 1000' in size_lower:
        return 'Medium'

    elif '501 to 1000' in size_lower or '1001 to 5000' in size_lower:
        return 'Large'

    elif '5001' in size_lower or '10000+' in size_lower:
        return 'Very Large'

    # If the size doesn't match any category
    else:
        return 'Unknown'


# Apply the function to the 'Size' column
# and create a new column with standardized categories
df['Company_Size_Category'] = df['Size'].apply(categorize_size)


# Display original size and the new category
print(df[['Size', 'Company_Size_Category']].head())
````

## 10. Extract Seniority Level
**(Difficulty: 🟢 Low - Feature Engineering)**

```
**Problem:**
Job titles often contain seniority information such as "Junior Data Analyst", "Senior Data Scientist", or "Lead Engineer". Extract the seniority level from the job title and categorize it into levels such as Junior, Senior, or Lead.

**Example:**

Input: "Senior Data Analyst"
Output: "Senior"

````
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/nisheshgogia/vorithm-interview-questions-solutions/main/Python/data_analytics/dataanalyst.csv")

def extract_seniority(job_title):
    if pd.isna(job_title):
        return 'Not Specified'

    title_lower = str(job_title).lower()

    if 'lead' in title_lower or 'principal' in title_lower or 'director' in title_lower:
        return 'Lead'
    elif 'senior' in title_lower or 'sr.' in title_lower or 'sr ' in title_lower:
        return 'Senior'
    elif 'junior' in title_lower or 'jr.' in title_lower or 'jr ' in title_lower:
        return 'Junior'
    elif 'entry' in title_lower or 'intern' in title_lower:
        return 'Entry Level'
    elif 'manager' in title_lower:
        return 'Manager'
    else:
        return 'Mid Level'

df['Seniority_Level'] = df['Job Title'].apply(extract_seniority)

print(df[['Job Title', 'Seniority_Level']].head(10))
print(f"\nSeniority Distribution:\n{df['Seniority_Level'].value_counts()}")


````

## 11. Extract Skill Indicators
**(Difficulty: 🟢 Low - Feature Engineering)**
```
Problem:
Job descriptions often mention required skills such as Python, SQL, Excel, or Tableau. Create binary features indicating whether each skill appears in the job description.

Example:
If description contains "Python and SQL", then python_flag=1 and sql_flag=1

````
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/nisheshgogia/vorithm-interview-questions-solutions/main/Python/data_analytics/dataanalyst.csv")

# Define skills to search for
skills = ['python', 'sql', 'excel', 'tableau', 'power bi', 'r ', 'sas',
          'java', 'spark', 'hadoop', 'aws', 'azure']

# Convert job description to lowercase for matching
df['Job Description'] = df['Job Description'].fillna('').str.lower()

# Create binary flag for each skill
for skill in skills:
    column_name = f'{skill.strip()}_flag'
    df[column_name] = np.where(df['Job Description'].str.contains(skill, na=False), 1, 0)

# Display results
skill_columns = [f'{skill.strip()}_flag' for skill in skills]
print(df[['Job Title'] + skill_columns].head())

# Summary statistics
print("\nSkill Frequency:")
for col in skill_columns:
    print(f"{col}: {df[col].sum()} jobs ({df[col].mean()*100:.1f}%)")

````

# 23. Detect Salary Outliers
# (Difficulty: 🟢 Low )
```
Problem: Detect unusually high or low salary values using the Interquartile Range (IQR) method.

Example: Identify salaries that fall outside the normal range of the dataset.

````
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/nisheshgogia/vorithm-interview-questions-solutions/main/Python/data_analytics/dataanalyst.csv")


# Step 1: Clean Salary Estimate column
# Remove extra text like "(Glassdoor est.)"
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Glassdoor est.)', '', regex=False)

# Remove rows where salary is missing
df = df[df['Salary Estimate'] != '-1']

# Step 2: Extract minimum and maximum salary
salary_split = df['Salary Estimate'].str.split('-', expand=True)

df['min_salary'] = salary_split[0]
df['max_salary'] = salary_split[1]

# Remove symbols ($ and K)
df['min_salary'] = df['min_salary'].str.replace('$', '', regex=False).str.replace('K', '', regex=False).str.strip()
df['max_salary'] = df['max_salary'].str.replace('$', '', regex=False).str.replace('K', '', regex=False).str.strip()

# Convert to numeric
df['min_salary'] = pd.to_numeric(df['min_salary'], errors='coerce') * 1000
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce') * 1000

# Step 3: Create average salary column
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2

# Step 4: Detect outliers using IQR method
Q1 = df['avg_salary'].quantile(0.25)
Q3 = df['avg_salary'].quantile(0.75)

IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 5: Flag outliers
df['Is_Outlier'] = np.where(
    (df['avg_salary'] < lower_bound) | (df['avg_salary'] > upper_bound),
    1,
    0
)
# Step 6: Display statistics

print(f"Q1: ${Q1:,.0f}")
print(f"Q3: ${Q3:,.0f}")
print(f"IQR: ${IQR:,.0f}")
print(f"Lower Bound: ${lower_bound:,.0f}")
print(f"Upper Bound: ${upper_bound:,.0f}")

print(f"\nTotal Outliers: {df['Is_Outlier'].sum()} ({df['Is_Outlier'].mean()*100:.1f}%)")

# Step 7: Show example outlier rows
outliers = df[df['Is_Outlier'] == 1][['Job Title', 'Company Name', 'avg_salary']]

print("\nSample Outliers:")
print(outliers.head(10))
````

````
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/nisheshgogia/vorithm-interview-questions-solutions/main/Python/data_analytics/dataanalyst.csv")

# Step 1: Clean Salary Estimate column

# Remove extra text like "(Glassdoor est.)"
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Glassdoor est.)', '', regex=False)

# Remove rows where salary is missing
df = df[df['Salary Estimate'] != '-1']

# Step 2: Extract minimum and maximum salary
salary_split = df['Salary Estimate'].str.split('-', expand=True)

df['min_salary'] = salary_split[0]
df['max_salary'] = salary_split[1]

# Remove symbols ($ and K)
df['min_salary'] = df['min_salary'].str.replace('$', '', regex=False).str.replace('K', '', regex=False).str.strip()
df['max_salary'] = df['max_salary'].str.replace('$', '', regex=False).str.replace('K', '', regex=False).str.strip()

# Convert to numeric
df['min_salary'] = pd.to_numeric(df['min_salary'], errors='coerce') * 1000
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce') * 1000

# Step 3: Create average salary column
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2

# Step 4: Calculate mean and standard deviation per industry
industry_stats = df.groupby('Industry')['avg_salary'].agg(['mean', 'std']).reset_index()

# Merge industry statistics back to the dataframe
df = df.merge(industry_stats, on='Industry', how='left')

# Step 5: Z-score normalization
# (value - mean) / standard deviation
df['Normalized_Salary'] = (df['avg_salary'] - df['mean']) / df['std']


# Step 6: Min-Max normalization
# (value - min) / (max - min)
industry_minmax = df.groupby('Industry')['avg_salary'].agg(['min', 'max']).reset_index()

# Merge min and max values
df = df.merge(industry_minmax, on='Industry', how='left')
df['Normalized_Salary_MinMax'] = (df['avg_salary'] - df['min']) / (df['max'] - df['min'])

# Step 7: Display results
print(df[['Industry', 'avg_salary', 'Normalized_Salary', 'Normalized_Salary_MinMax']].head(10))

# Remove temporary columns
df = df.drop(columns=['mean', 'std', 'min', 'max'])
````
