# 1. Delete a Node in Linked List (Given Only That Node)
# Difficulty: 🟢 LowProblem Statement:

There is a singly-linked list head and we want to delete a node in it. You are given the node to be deleted. You will not be given access to the first node of head.All the values of the linked list are unique, and it is guaranteed that the given node is not the last node in the linked list.Delete the given node. Note that by deleting the node, we mean removing it from the linked list, not just setting its value to null.
```
Example 1:
Input: head = [4,5,1,9], node = 5
Output: [4,1,9]
Explanation: You are given node with value 5. After deleting it, the list becomes 4 → 1 → 9
```
Example 2:

Input: head = [4,5,1,9], node = 1
Output: [4,5,9]
```
Constraints:
The number of nodes in the list is in the range [2, 1000]
The given node is not the tail
All node values are unique

````python
# Definition for singly-linked list node
class ListNode:
    def __init__(self, val=0, next=None):
        # 'val' stores the value of the node
        self.val = val

        # 'next' stores the reference to the next node
        self.next = next


# Function to delete a node when only that node is given
def delete_node_brute(node):
    """
    Since we do NOT have access to the previous node,
    we cannot directly remove this node.

    Trick:
    1. Copy the value of the next node into this node
    2. Skip the next node

    This effectively removes the next node, but the current
    node now contains the next node's value.
    """

    # If node is None OR it is the last node, deletion is impossible
    if node is None or node.next is None:
        return

    # Copy next node's value into current node
    node.val = node.next.val

    # Skip the next node (remove it from the list)
    node.next = node.next.next


# Function to print the linked list
def print_list(head):
    """
    Traverse the linked list and print all node values
    """
    current = head

    # Loop until we reach the end of the list
    while current:
        print(current.val, end=" -> ")
        current = current.next

    # Indicate the end of the list
    print("None")


# -------------------------
# Creating the Linked List
# -------------------------

# Create nodes manually
head = ListNode(4)
head.next = ListNode(5)
head.next.next = ListNode(1)
head.next.next.next = ListNode(9)

# Linked list now looks like:
# 4 -> 5 -> 1 -> 9 -> None


print("Before deletion:")
print_list(head)


# Delete the node with value 5
# We pass the node itself (head.next)
delete_node_brute(head.next)


print("After deletion:")
print_list(head)


# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1) : only three operations: copy value, update pointer
#                no loops or recursion involved
# avg   → O(1) : always exactly 2 assignments regardless of list size
#                node.val = node.next.val → O(1)
#                node.next = node.next.next → O(1)
# worst → O(1) : even in large list, only local operations performed
#                doesn't traverse the list at all
# space → O(1) : no extra data structures, only pointer manipulation

#best time complexity    - O(1)
#average time complexity - O(1)
#worst time complexity   - O(1)
#space complexity        - O(1)
````

````python
#Optimized Approach (Same as Brute Force - Already Optimal)
def delete_node_optimized(node):
    """
    This is already optimal. There's no way to do better than O(1)
    since we can't access the previous node.
    """
    # Copy next node's value
    node.val = node.next.val

    # Bypass next node
    node.next = node.next.next


# -------- Added only to print output --------

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1) : constant time - two pointer operations
# avg   → O(1) : same operations regardless of input
# worst → O(1) : cannot be improved further
# space → O(1) : in-place modification

# IMPORTANT NOTE: This is a "trick" solution
# We're not actually deleting the given node - we're copying
# the next node's data and deleting the next node instead.
# This works because the problem guarantees the node is not the tail.

#best time complexity    - O(1)
#average time complexity - O(1)
#worst time complexity   - O(1)
#space complexity        - O(1)


````


## 2. Reverse a Linked List
**Difficulty: 🟢 Low**

**Problem Statement:**
Given the head of a singly linked list, reverse the list, and return the reversed list's new head.

**Example 1:**
```
Input: head = [1,2,3,4,5]
Output: [5,4,3,2,1]
Explanation: 1 → 2 → 3 → 4 → 5 becomes 5 → 4 → 3 → 2 → 1
```

**Example 2:**
```
Input: head = [1,2]
Output: [2,1]
```

**Example 3:**
```
Input: head = []
Output: []

````python
#Brute Force Approach (Store in Array, Rebuild)
def reverse_list_brute(head):
    if not head:
        return None

    # Step 1: Store all values in array
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next

    # Step 2: Reverse array
    values.reverse()

    # Step 3: Create new linked list from reversed array
    dummy = ListNode(0)
    current = dummy
    for val in values:
        current.next = ListNode(val)
        current = current.next

    return dummy.next


# Needed so code runs
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


# Example
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

reversed_head = reverse_list_brute(head)

# -------- Added only to print output --------
current = reversed_head
while current:
    print(current.val, end=" -> ")
    current = current.next
print("None")


# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : empty list returns immediately O(1)
#                non-empty list: must traverse all n nodes to store values
# avg   → O(n) : step 1: traverse n nodes → O(n)
#                step 2: reverse array of n elements → O(n)
#                step 3: create n new nodes → O(n)
#                total = 3n = O(n)
# worst → O(n) : large list still only requires one pass for each step
# space → O(n) : values array stores n elements
#                new linked list also uses n nodes (even though it's output)

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

````python
# Needed so the code runs
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


#Optimized Approach (Iterative In-Place Reversal)
def reverse_list_optimized(head):
    prev = None
    current = head

    while current:
        next_temp = current.next    # Save next node
        current.next = prev         # Reverse pointer
        prev = current              # Move prev forward
        current = next_temp         # Move current forward

    return prev  # prev is now the new head


# Example
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
reversed_head = reverse_list_optimized(head)

# -------- Added only to print output --------
curr = reversed_head
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")


# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : even single node requires loop entry check
#                empty list returns immediately but still O(1) ≈ O(n) for n=0
# avg   → O(n) : while loop iterates exactly n times (once per node)
#                each iteration: 3 pointer assignments = O(1)
#                total = n × O(1) = O(n)
# worst → O(n) : large list still single pass with constant work per node
# space → O(1) : only three pointers (prev, current, next_temp)
#                reversal done in-place by changing next pointers

# VISUALIZATION:
# Initial:  1 → 2 → 3 → 4 → 5 → None
#          prev curr
# Step 1:   None ← 1   2 → 3 → 4 → 5 → None
#                prev curr
# Step 2:   None ← 1 ← 2   3 → 4 → 5 → None
#                     prev curr
# Final:    None ← 1 ← 2 ← 3 ← 4 ← 5
#                                  prev (return this)

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

## 3. Detect Cycle in Linked List
**Difficulty: 🟡 Medium**

**Problem Statement:**
Given head, the head of a linked list, determine if the linked list has a cycle in it.

A cycle exists if there is some node in the list that can be reached again by continuously following the next pointer. Return true if there is a cycle, otherwise return false.

**Example 1:**
```
Input: head = [3,2,0,-4], pos = 1
Output: true
Explanation: There is a cycle where the tail connects to the 1st node (0-indexed)
Visual: 3 → 2 → 0 → -4
            ↑_________|
```

**Example 2:**
```
Input: head = [1,2], pos = 0
Output: true
Explanation: Cycle connects tail to head
```

**Example 3:**
```
Input: head = [1], pos = -1
Output: false
Explanation: No cycle

````python
#Brute Force Approach (HashSet to Track Visited Nodes)
def has_cycle_brute(head):
    visited = set()
    current = head

    while current:
        if current in visited:
            return True         # Found a node we've seen before
        visited.add(current)    # Mark as visited
        current = current.next

    return False  # Reached end without finding cycle

# Example with cycle
head = ListNode(3)
second = ListNode(2)
third = ListNode(0)
fourth = ListNode(-4)
head.next = second
second.next = third
third.next = fourth
fourth.next = second  # Cycle back to second node

print(has_cycle_brute(head))  # Output: True

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1) : cycle at head (head.next points to head itself)
#                detected on second iteration → O(1)
# avg   → O(n) : if cycle exists at position k, we visit k+1 nodes
#                if no cycle, we visit all n nodes → O(n)
#                average case visits ~n/2 nodes
# worst → O(n) : no cycle → traverse entire list of n nodes
#                or cycle at very end → visit all nodes before detecting
# space → O(n) : visited set stores up to n node references
#                in worst case (no cycle), all n nodes stored

#best time complexity    - O(1)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

````python
#Optimized Approach (Floyd's Cycle Detection - Slow/Fast Pointers)
def has_cycle_optimized(head):
    if not head or not head.next:
        return False

    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next        # Move 1 step
        fast = fast.next.next   # Move 2 steps

        if slow == fast:        # They met - cycle exists
            return True

    return False  # Fast reached end - no cycle

print(has_cycle_optimized(head))  # Output: True

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1) : empty list or single node → immediate return
# avg   → O(n) : if cycle exists, slow/fast meet after at most n steps
#                mathematical proof: if cycle length is C and list length is n,
#                they meet after at most n iterations
# worst → O(n) : no cycle → fast reaches end after n/2 iterations
#                with cycle → slow enters cycle, fast catches up in ≤n steps
#                total distance traveled ≤ 2n (fast moves twice as fast)
# space → O(1) : only two pointers (slow, fast), no extra data structures

# WHY IT WORKS (Floyd's Algorithm):
# If there's a cycle, the fast pointer will eventually "lap" the slow pointer
# Think of it like two runners on a circular track - the faster one will
# eventually catch up to the slower one from behind

# VISUALIZATION:
# 1 → 2 → 3 → 4 → 5
#         ↑_______|
# Step 1: slow=1, fast=1
# Step 2: slow=2, fast=3
# Step 3: slow=3, fast=5
# Step 4: slow=4, fast=4 (they meet!)

#best time complexity    - O(1)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)


````

## 4. Find Middle of Linked List
**Difficulty: 🟢 Low**

**Problem Statement:**
Given the head of a singly linked list, return the middle node of the linked list.

If there are two middle nodes, return the second middle node.

**Example 1:**
```
Input: head = [1,2,3,4,5]
Output: [3,4,5]
Explanation: The middle node is 3 (third node)
```

**Example 2:**
```
Input: head = [1,2,3,4,5,6]
Output: [4,5,6]
Explanation: Two middle nodes (3 and 4), return second one (4)

````python
#Brute Force Approach (Count Length, Then Find Middle)
def middle_node_brute(head):
    if not head:
        return None

    # Step 1: Count total nodes
    length = 0
    current = head
    while current:
        length += 1
        current = current.next

    # Step 2: Find middle position
    middle_pos = length // 2

    # Step 3: Traverse to middle
    current = head
    for i in range(middle_pos):
        current = current.next

    return current

# Example
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
middle = middle_node_brute(head)
print(middle.val)  # Output: 3

head2 = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5, ListNode(6))))))
middle2 = middle_node_brute(head2)
print(middle2.val)  # Output: 4

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : even single node requires counting (length=1)
#                then no traversal needed but still visited once
# avg   → O(n) : step 1: traverse n nodes to count → O(n)
#                step 2: calculation → O(1)
#                step 3: traverse n/2 nodes to middle → O(n/2)
#                total = n + n/2 = 1.5n = O(n)
# worst → O(n) : large list still requires full count pass + half traversal
# space → O(1) : only length and current pointer variables

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

````python
#Optimized Approach (Slow/Fast Pointer - Tortoise and Hare)
def middle_node_optimized(head):
    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next        # Move 1 step
        fast = fast.next.next   # Move 2 steps

    return slow  # When fast reaches end, slow is at middle

# Example
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
middle = middle_node_optimized(head)
print(middle.val)  # Output: 3

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : even for 1-2 nodes, loop condition checked
#                empty list returns immediately but n=0
# avg   → O(n) : while loop runs n/2 times (fast moves 2× speed)
#                slow moves n/2 steps, fast moves n steps
#                total pointer movements = n/2 + n = 1.5n but still O(n)
# worst → O(n) : large list: fast traverses full n nodes (in n/2 steps)
#                single pass algorithm - more efficient than two passes
# space → O(1) : only two pointers (slow, fast)

# WHY IT WORKS:
# When fast (moving 2x speed) reaches end, slow is at exactly middle
# For odd length: fast lands on last node
# For even length: fast lands on None, slow on second middle

# VISUALIZATION (odd length n=5):
# Initial:  1 → 2 → 3 → 4 → 5
#          slow,fast
# Step 1:   1 → 2 → 3 → 4 → 5
#              slow   fast
# Step 2:   1 → 2 → 3 → 4 → 5
#                  slow     fast (end)
# Return slow (node 3)

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)

````

## 5. Merge Two Sorted Linked Lists
**Difficulty: 🟢 Low**

**Problem Statement:**
Merge two sorted linked lists and return it as a sorted list. The list should be made by splicing together the nodes of the first two lists.

**Example 1:**
```
Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
```

**Example 2:**
```
Input: list1 = [], list2 = []
Output: []
```

**Example 3:**
```
Input: list1 = [], list2 = [0]
Output: [0]

````python
#Brute Force Approach (Collect All Values, Sort, Rebuild)
def merge_two_lists_brute(list1, list2):
    # Step 1: Collect all values
    values = []
    current = list1
    while current:
        values.append(current.val)
        current = current.next

    current = list2
    while current:
        values.append(current.val)
        current = current.next

    # Step 2: Sort values
    values.sort()

    # Step 3: Build new linked list
    dummy = ListNode(0)
    current = dummy
    for val in values:
        current.next = ListNode(val)
        current = current.next

    return dummy.next

# Example
list1 = ListNode(1, ListNode(2, ListNode(4)))
list2 = ListNode(1, ListNode(3, ListNode(4)))
merged = merge_two_lists_brute(list1, list2)
# Result: 1 → 1 → 2 → 3 → 4 → 4

# HOW COMPLEXITIES ARE DERIVED:
# best  → O((m+n)log(m+n)) : even if lists sorted, still need to:
#                            collect: O(m+n), sort: O((m+n)log(m+n))
#                            cannot skip sorting step
# avg   → O((m+n)log(m+n)) : step 1: traverse m+n nodes → O(m+n)
#                            step 2: sort (m+n) values → O((m+n)log(m+n))
#                            step 3: create (m+n) nodes → O(m+n)
#                            dominated by sorting: O((m+n)log(m+n))
# worst → O((m+n)log(m+n)) : large lists still require full sort
# space → O(m+n) : values array stores m+n elements

#best time complexity    - O((m+n)log(m+n))
#average time complexity - O((m+n)log(m+n))
#worst time complexity   - O((m+n)log(m+n))
#space complexity        - O(m+n)
````

````python
# Needed so the code runs
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


#Optimized Approach (Two Pointer Merge - Like Merge Sort)
def merge_two_lists_optimized(list1, list2):
    dummy = ListNode(0)
    current = dummy

    while list1 and list2:
        if list1.val <= list2.val:
            current.next = list1
            list1 = list1.next
        else:
            current.next = list2
            list2 = list2.next
        current = current.next

    # Attach remaining nodes (one list is exhausted)
    if list1:
        current.next = list1
    if list2:
        current.next = list2

    return dummy.next


# Example
list1 = ListNode(1, ListNode(2, ListNode(4)))
list2 = ListNode(1, ListNode(3, ListNode(4)))

merged = merge_two_lists_optimized(list1, list2)

# -------- Added only to print output --------
curr = merged
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")


# HOW COMPLEXITIES ARE DERIVED:
# best  → O(m+n) : even if all of list1 < list2, still attach all nodes
#                  must visit every node from both lists once
# avg   → O(m+n) : while loop runs min(m,n) times
#                  each iteration: one comparison + one pointer update = O(1)
#                  after loop: attach remaining max(m,n)-min(m,n) nodes
#                  total = min(m,n) + (max(m,n)-min(m,n)) = max(m,n) = O(m+n)
# worst → O(m+n) : alternating values → while runs min(m,n) times
#                  example: [1,3,5] + [2,4,6] → all nodes processed in loop
# space → O(1)   : only dummy and current pointers
#                  reusing existing nodes, not creating new ones

# KEY INSIGHT: Lists already sorted, so we can merge in single pass
# Similar to merge step in merge sort algorithm

#best time complexity    - O(m+n)
#average time complexity - O(m+n)
#worst time complexity   - O(m+n)
#space complexity        - O(1)
````

## 6. Remove Nth Node From End
**Difficulty: 🟡 Medium**

**Problem Statement:**
Given the head of a linked list, remove the n-th node from the end of the list and return its head.

**Example 1:**
```
Input: head = [1,2,3,4,5], n = 2
Output: [1,2,3,5]
Explanation: Remove 4 (2nd node from end)
1 → 2 → 3 → 4 → 5
Remove 4: 1 → 2 → 3 → 5
```

**Example 2:**
```
Input: head = [1], n = 1
Output: []
Explanation: Only one node, remove it
```

**Example 3:**
```
Input: head = [1,2], n = 1
Output: [1]
Explanation: Remove last node

````python
# Needed so the code runs
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


#Brute Force Approach (Count Length, Calculate Position, Delete)
def remove_nth_from_end_brute(head, n):
    # Step 1: Count total nodes
    length = 0
    current = head
    while current:
        length += 1
        current = current.next

    # Step 2: Calculate position from beginning
    position = length - n

    # Step 3: Handle edge case - removing head
    if position == 0:
        return head.next

    # Step 4: Traverse to node before target
    current = head
    for i in range(position - 1):
        current = current.next

    # Step 5: Remove target node
    current.next = current.next.next

    return head


# Example
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
result = remove_nth_from_end_brute(head, 2)

# -------- Added only to print output --------
curr = result
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")


# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : removing first node (n=length) requires counting all nodes
#                position calculation immediate but still counted O(n)
# avg   → O(n) : step 1: count n nodes → O(n)
#                step 4: traverse (n-k) nodes → O(n-k) where k is position from end
#                total = n + (n-k) < 2n = O(n)
# worst → O(n) : removing last node requires count O(n) + traverse (n-1) O(n)
#                total = 2n = O(n)
# space → O(1) : only length, position, current pointer variables

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

````python
# Needed so the code runs
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


#Optimized Approach (Two Pointers with Gap - Single Pass)
def remove_nth_from_end_optimized(head, n):
    dummy = ListNode(0)
    dummy.next = head
    first = dummy
    second = dummy

    # Step 1: Move first pointer n+1 steps ahead
    for i in range(n + 1):
        first = first.next

    # Step 2: Move both pointers until first reaches end
    # This maintains gap of n between them
    while first:
        first = first.next
        second = second.next

    # Step 3: Remove the nth node from end
    second.next = second.next.next

    return dummy.next


# Example
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
result = remove_nth_from_end_optimized(head, 2)

# -------- Added only to print output --------
curr = result
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")


# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : even removing first node, must move first pointer n steps
#                then move both pointers to end
# avg   → O(n) : step 1: move first n+1 steps → O(n)
#                step 2: move both pointers (length-n-1) steps
#                total pointer movements ≈ length = O(n)
#                single pass through list
# worst → O(n) : removing last node still only one pass
#                total movements = n+1 + (length-n-1) = length = O(n)
# space → O(1) : only dummy, first, second pointers

# WHY IT WORKS (Two Pointer Technique):
# Maintain gap of n between pointers
# When first reaches end, second is at node before target

# VISUALIZATION (n=2):
# Initial:  D → 1 → 2 → 3 → 4 → 5 → None
#          1st,2nd
# After gap of n+1:
#          D → 1 → 2 → 3 → 4 → 5 → None
#          2nd         1st
# Move both together:
#          D → 1 → 2 → 3 → 4 → 5 → None
#                      2nd         1st
# Now second.next is the node to remove (4)

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````
