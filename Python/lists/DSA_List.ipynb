# 1. Remove Duplicates While Preserving Order (Difficulty: 🟢 Low)
Given a list of integers nums, remove duplicates while keeping the original order.

Examples

```
Example 1
Input: nums = [1,2,2,3,1]
Output: [1,2,3]
```
```
Example 2
Input: nums = [4,4,5,6,5]
Output: [4,5,6]

````python
#Brute Force Approach
def remove_duplicates_brute(nums):
    result = []
    for num in nums:       # loop runs n times
        if num not in result:  # 'in' scans entire result list → O(n) each time
            result.append(num)
    return result

print(remove_duplicates_brute([1,2,2,3,1]))  # Output: [1,2,3]
print(remove_duplicates_brute([4,4,5,6,5]))  # Output: [4,5,6]


# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : all elements are unique, 'in' check always finds nothing quickly
#                  but we still scan the full result — actually still O(n^2) in theory,
#                  but with tiny result lists early on it behaves closer to O(n)
# avg   → O(n^2) : for each of n elements, 'in result' scans up to n elements
#                  → n * n = O(n^2)
# worst → O(n^2) : all elements unique → result grows to size n,
#                  last element check scans all n → n*(n+1)/2 comparisons = O(n^2)
# space → O(n)   : result list stores up to n unique elements


#best time complexity    - O(n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(n)
````

````python
#Optimized Approach (LinkedHashSet via dict)
def remove_duplicates_optimized(nums):
    seen = {}        # dict gives O(1) average lookup
    result = []
    for num in nums: # single pass → O(n)
        if num not in seen:   # O(1) hash lookup
            seen[num] = True
            result.append(num)
    return result

print(remove_duplicates_optimized([1,2,2,3,1]))  # Output: [1,2,3]
print(remove_duplicates_optimized([4,4,5,6,5]))  # Output: [4,5,6]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : single pass through n elements, each hash lookup is O(1)
# avg   → O(n) : hash dict gives O(1) average for all operations, loop runs n times
# worst → O(n) : even with hash collisions Python rehashes internally, still O(n)
# space → O(n) : seen dict + result list each hold up to n elements

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

# 2. Remove All Occurrences of a Value (Difficulty: 🟢 Low)
Given a list nums and an integer val, remove all occurrences of val.

Examples
```
Example 1

Input: nums = [3,2,2,3], val = 3
Output: [2,2]
```
```
Example 2

Input: nums = [1,1,2,3], val = 1
Output: [2,3]

````python
#Brute Force Approach
def remove_value_brute(nums, val):
    result = []
    for num in nums:      # single pass through n elements
        if num != val:    # simple comparison O(1)
            result.append(num)
    return result

print(remove_value_brute([3,2,2,3], 3))  # Output: [2,2]
print(remove_value_brute([1,1,2,3], 1))  # Output: [2,3]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : even if val not present, we still scan all n elements
# avg   → O(n) : always one full pass regardless of how many val's exist
# worst → O(n) : all elements equal val → scan all n, append nothing still O(n)
# space → O(n) : result list holds up to n elements in worst case (no val found)

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

````python
#Optimized Approach (Two Pointer In-Place)
def remove_value_optimized(nums, val):
    write = 0                       # write pointer starts at 0
    for read in range(len(nums)):   # read pointer scans all n elements
        if nums[read] != val:       # O(1) comparison
            nums[write] = nums[read]# overwrite in-place
            write += 1
    return nums[:write]

print(remove_value_optimized([3,2,2,3], 3))  # Output: [2,2]
print(remove_value_optimized([1,1,2,3], 1))  # Output: [2,3]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : read pointer always traverses all n elements (no early exit)
# avg   → O(n) : single pass with two pointers, each moving at most n steps
# worst → O(n) : no val found → write = read for all n, still one full pass
# space → O(1) : modification done in-place, only two integer pointers used

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

# 3. Intersection of Two Lists (Difficulty: 🟡 Medium
)
Given two lists, return a list of common elements.

Examples
```
Example 1

Input: a = [1,2,2,3], b = [2,2]
Output: [2,2]
```
```
Example 2

Input: a = [4,5,6], b = [7,8]
Output: []

````python
#Brute Force Approach
def intersection_brute(a, b):
    result = []
    b_copy = b[:]           # copy b to allow removal
    for num in a:           # outer loop: n iterations
        if num in b_copy:   # inner scan of b_copy: O(m) each time
            result.append(num)
            b_copy.remove(num)  # remove also O(m)
    return result

print(intersection_brute([1,2,2,3], [2,2]))  # Output: [2,2]
print(intersection_brute([4,5,6], [7,8]))    # Output: []

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : no common elements → 'in b_copy' fails fast each time
#                  but still checks all of b_copy → still O(n*m)
# avg   → O(n*m) : for each of n elements in a, scan up to m elements in b_copy
# worst → O(n*m) : all elements match → n elements × m scan + m removal = O(n*m)
# space → O(m)   : b_copy stores m elements + result stores min(n,m) elements

#best time complexity    - O(n)
#average time complexity - O(n * m)
#worst time complexity   - O(n * m)
#space complexity        - O(m)
````

````python
#Optimized Approach (HashMap Frequency Count)
from collections import Counter

def intersection_optimized(a, b):
    count_a = Counter(a)  # O(n) to build
    count_b = Counter(b)  # O(m) to build
    result = []
    for num in count_a:                            # iterate unique keys of a
        if num in count_b:                         # O(1) hash lookup
            result.extend([num] * min(count_a[num], count_b[num]))
    return result

print(intersection_optimized([1,2,2,3], [2,2]))  # Output: [2,2]
print(intersection_optimized([4,5,6], [7,8]))    # Output: []

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n+m) : building both counters takes O(n) + O(m),
#                  loop over unique keys is at most O(n), total O(n+m)
# avg   → O(n+m) : Counter builds in one pass each, lookup is O(1)
# worst → O(n+m) : all elements unique → counters still built in O(n+m)
# space → O(n+m) : count_a holds n entries, count_b holds m entries

#best time complexity    - O(n + m)
#average time complexity - O(n + m)
#worst time complexity   - O(n + m)
#space complexity        - O(n + m)
````

# 4.Most Frequent Element

Given an array of integers, return the element that appears most frequently.

Examples
```
Example 1

Input: nums = [1,1,2,3,3,3]
Output: 3
```
```
Example 2

Input: nums = [4,4,4,2,2]
Output: 4

````python
def most_frequent_brute(nums):
    max_count = 0
    result = nums[0]
    for i in range(len(nums)):        # outer loop: n iterations
        count = 0
        for j in range(len(nums)):    # inner loop: n iterations for each i
            if nums[j] == nums[i]:
                count += 1
        if count > max_count:
            max_count = count
            result = nums[i]
    return result

print(most_frequent_brute([1,1,2,3,3,3]))  # Output: 3
print(most_frequent_brute([4,4,4,2,2]))    # Output: 4

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : if first element is the most frequent AND we detect early
#                  but since inner loop always runs n times → still O(n^2)
# avg   → O(n^2) : outer loop n × inner loop n = n^2 comparisons always
# worst → O(n^2) : all elements unique → still runs full n × n loops
# space → O(1)   : only two variables (max_count, result), no extra structure

#best time complexity    - O(n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python
#Optimized Approach (HashMap Frequency Count)
def most_frequent_optimized(nums):
    count = {}
    for num in nums:                       # single pass O(n)
        count[num] = count.get(num, 0) + 1 # O(1) hash update
    return max(count, key=count.get)       # O(k) where k = unique elements ≤ n

print(most_frequent_optimized([1,1,2,3,3,3]))  # Output: 3
print(most_frequent_optimized([4,4,4,2,2]))    # Output: 4

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : single pass to build frequency dict + one pass for max = O(n)
# avg   → O(n) : hash operations are O(1) average, loop runs n times
# worst → O(n) : all unique elements → dict has n keys, max scan is O(n)
# space → O(n) : dict stores up to n key-value pairs in worst case

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

# 5. Rotate a List by K Positions (Difficulty: 🟡 Medium)
Given a list nums and integer k, rotate the list to the right by k steps.

Examples
```
Example 1

Input: nums = [1,2,3,4,5], k = 2
Output: [4,5,1,2,3]
```
```
Example 2

Input: nums = [10,20,30], k = 1
Output: [30,10,20]


````python
#Brute Force Approach
def rotate_brute(nums, k):
    k = k % len(nums)           # normalize k
    for _ in range(k):          # outer loop: k iterations
        last = nums[-1]
        for i in range(len(nums) - 1, 0, -1):  # inner loop: n-1 shifts
            nums[i] = nums[i - 1]
        nums[0] = last
    return nums

print(rotate_brute([1,2,3,4,5], 2))  # Output: [4,5,1,2,3]
print(rotate_brute([10,20,30], 1))   # Output: [30,10,20]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : k=1 after modulo → inner loop runs once for n-1 shifts
# avg   → O(n*k) : outer loop k times × inner loop n shifts = n*k operations
# worst → O(n*k) : k = n/2 (max useful rotation) → n * n/2 = O(n^2) effectively
# space → O(1)   : rotation done in-place with single temp variable

#best time complexity    - O(n)
#average time complexity - O(n * k)
#worst time complexity   - O(n * k)
#space complexity        - O(1)
````

````python
#Optimized Approach (Slice and Concatenate)
def rotate_optimized(nums, k):
    k = k % len(nums)         # normalize: handles k > len(nums)
    return nums[-k:] + nums[:-k]
    # nums[-k:]  → last k elements   (the part that wraps to front)
    # nums[:-k]  → first n-k elements (the part that shifts right)

print(rotate_optimized([1,2,3,4,5], 2))  # Output: [4,5,1,2,3]
print(rotate_optimized([10,20,30], 1))   # Output: [30,10,20]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : slicing nums[-k:] takes O(k), nums[:-k] takes O(n-k),
#                concatenation takes O(n) → total O(n) regardless of k
# avg   → O(n) : slice + concat always proportional to total list length n
# worst → O(n) : k = n//2 → both slices together still cover n elements
# space → O(n) : two slices + concatenated result each hold parts of n elements

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

# 6. Two Sum (Find Indices) Difficulty: 🟢 Low
Given an array and a target, return indices of two numbers that add up to the target.

Examples
```
Example 1

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
```
```
Example 2

Input: nums = [3,2,4], target = 6
Output: [1,2]

````python
#Brute Force Approach
def two_sum_brute(nums, target):
    for i in range(len(nums)):          # outer loop: n iterations
        for j in range(i + 1, len(nums)):# inner loop: n-i-1 iterations
            if nums[i] + nums[j] == target:
                return [i, j]           # early exit on first found pair
    return []

print(two_sum_brute([2,7,11,15], 9))  # Output: [0,1]
print(two_sum_brute([3,2,4], 6))      # Output: [1,2]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1)   : answer is first pair (i=0, j=1) → found in 1 comparison
# avg   → O(n^2) : pair found somewhere in the middle → ~n^2/2 comparisons
# worst → O(n^2) : pair at end or doesn't exist → n*(n-1)/2 comparisons = O(n^2)
# space → O(1)   : no extra data structures, just two index variables

#best time complexity    - O(1)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python
#Optimized Approach (HashMap)
def two_sum_optimized(nums, target):
    seen = {}                           # stores {value: index}
    for i, num in enumerate(nums):      # single pass O(n)
        complement = target - num       # what we need to find
        if complement in seen:          # O(1) hash lookup
            return [seen[complement], i]
        seen[num] = i                   # store current value and index
    return []

print(two_sum_optimized([2,7,11,15], 9))  # Output: [0,1]
print(two_sum_optimized([3,2,4], 6))      # Output: [1,2]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1)   : complement found on second element (i=1) → 1 lookup
# avg   → O(n)   : complement found somewhere mid-array → ~n/2 iterations
# worst → O(n)   : pair at very end or no pair → full n iterations
# space → O(n)   : seen dict stores up to n-1 elements before finding answer

#best time complexity    - O(1)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

# 7. Merge Two Sorted Lists (Difficulty: 🟢 Low)

````python
#Brute Force Approach
def merge_sorted_brute(a, b):
    merged = a + b                      # combine both lists O(n+m)
    for i in range(len(merged)):        # outer loop: n+m iterations
        for j in range(i + 1, len(merged)): # inner loop: up to n+m
            if merged[i] > merged[j]:
                merged[i], merged[j] = merged[j], merged[i]  # bubble swap
    return merged

print(merge_sorted_brute([1,3,5], [2,4,6]))  # Output: [1,2,3,4,5,6]
print(merge_sorted_brute([1,2], [3,4]))      # Output: [1,2,3,4]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n log n) : if using Python's sort internally (timsort),
#                      but bubble sort here is O((n+m)^2) even best case
# avg   → O(n^2)     : bubble sort on n+m elements → (n+m)^2 / 2 comparisons
# worst → O(n^2)     : reverse sorted input → max swaps every iteration
# space → O(n)       : merged list holds all n+m elements

#best time complexity    - O(n log n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(n)
````

````python
#Optimized Approach (Two Pointer Merge)
def merge_sorted_optimized(a, b):
    result = []
    i = j = 0
    while i < len(a) and j < len(b):  # runs min(n,m) times
        if a[i] <= b[j]:              # O(1) comparison
            result.append(a[i])
            i += 1
        else:
            result.append(b[j])
            j += 1
    result.extend(a[i:])  # append leftovers O(n-i)
    result.extend(b[j:])  # append leftovers O(m-j)
    return result

print(merge_sorted_optimized([1,3,5], [2,4,6]))  # Output: [1,2,3,4,5,6]
print(merge_sorted_optimized([1,2], [3,4]))      # Output: [1,2,3,4]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)     : all elements of a < all elements of b →
#                    while loop runs n times, then extend b in one shot
# avg   → O(n+m)   : pointers traverse both lists once → total n+m steps
# worst → O(n+m)   : elements perfectly interleaved → while runs n+m-1 times
# space → O(n+m)   : result list stores all n+m elements

#best time complexity    - O(n)
#average time complexity - O(n + m)
#worst time complexity   - O(n + m)
#space complexity        - O(n + m)
````

# 8. Max Profit from Stock Prices (Difficulty: 🟡 Medium)
Return the maximum profit by buying and selling once.

Examples
```
Example 1

Input: prices = [7,1,5,3,6,4]
Output: 5
```
```
Example 2

Input: prices = [7,6,4,3,1]
Output: 0

````python
#Brute Force Approach
def max_profit_brute(prices):
    max_profit = 0
    for i in range(len(prices)):           # outer loop: n iterations (buy day)
        for j in range(i + 1, len(prices)):# inner loop: n-i-1 iterations (sell day)
            profit = prices[j] - prices[i]
            if profit > max_profit:
                max_profit = profit
    return max_profit

print(max_profit_brute([7,1,5,3,6,4]))  # Output: 5
print(max_profit_brute([7,6,4,3,1]))    # Output: 0

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : prices always descending → profit always ≤ 0,
#                  but inner loop still runs n-i-1 times → still O(n^2)
# avg   → O(n^2) : every (i,j) pair checked → n*(n-1)/2 comparisons = O(n^2)
# worst → O(n^2) : best profit at last pair → scan everything before returning
# space → O(1)   : only max_profit variable stored, no extra structure

#best time complexity    - O(n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python
#Optimized Approach (Single Pass Greedy)
def max_profit_optimized(prices):
    min_price = float('inf')   # track the lowest buy price seen so far
    max_profit = 0
    for price in prices:       # single pass O(n)
        if price < min_price:
            min_price = price  # update cheapest buy day
        elif price - min_price > max_profit:
            max_profit = price - min_price  # update best profit
    return max_profit

print(max_profit_optimized([7,1,5,3,6,4]))  # Output: 5
print(max_profit_optimized([7,6,4,3,1]))    # Output: 0

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : even if profit found early, we scan all n to ensure it's max
# avg   → O(n) : one pass, two comparisons per element → 2n operations = O(n)
# worst → O(n) : prices strictly decreasing → min updates every step, still O(n)
# space → O(1) : only two variables (min_price, max_profit) regardless of input

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

# 9. Euclidean Distance Between Two Lists (Difficulty: 🟢 Low)

Given two equal-length lists representing points, return the Euclidean distance.

Examples
```

Example 1

Input: a = [1,2], b = [4,6]
Output: 5.0
```
```
Example 2

Input: a = [0,0], b = [3,4]
Output: 5.0

````python
#Brute Force Approach
def euclidean_brute(a, b):
    total = 0
    for i in range(len(a)):   # single loop over n dimensions
        diff = a[i] - b[i]    # O(1) subtraction
        total += diff * diff  # O(1) multiply and add
    result = total ** 0.5     # O(1) square root
    return result

print(euclidean_brute([1,2], [4,6]))  # Output: 5.0
print(euclidean_brute([0,0], [3,4]))  # Output: 5.0

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : always loops through all n dimensions, no early exit possible
# avg   → O(n) : each dimension requires constant work → n * O(1) = O(n)
# worst → O(n) : regardless of values, all n pairs must be processed
# space → O(1) : only 'total' and 'diff' scalar variables used

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

````python
#Optimized Approach (zip + sum + math.sqrt)
import math

def euclidean_optimized(a, b):
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(a, b)))
    # zip(a,b)     → pairs elements lazily O(1) memory
    # generator    → computes (x-y)^2 on the fly, no intermediate list
    # sum()        → accumulates in single pass O(n)
    # math.sqrt()  → O(1) final operation

print(euclidean_optimized([1,2], [4,6]))  # Output: 5.0
print(euclidean_optimized([0,0], [3,4]))  # Output: 5.0

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : generator + sum is one pass through n pairs
# avg   → O(n) : zip lazily pairs, generator computes on demand, all O(1) per step
# worst → O(n) : all n dimensions always processed → n iterations
# space → O(1) : generator expression avoids storing intermediate list,
#                zip pairs one at a time → constant extra memory

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

# 10. Running Sum of a List (Difficulty: 🟢 Low)

Return the running sum of the list.

Examples
```
Example 1

Input: nums = [1,2,3,4]
Output: [1,3,6,10]
```
```
Example 2

Input: nums = [5,1,2]
Output: [5,6,8]

````python
#Brute Force Approach
def running_sum_brute(nums):
    result = []
    for i in range(len(nums)):      # outer loop: n iterations
        total = 0
        for j in range(i + 1):     # inner loop: i+1 iterations (recomputes sum)
            total += nums[j]
        result.append(total)
    return result

print(running_sum_brute([1,2,3,4]))  # Output: [1,3,6,10]
print(running_sum_brute([5,1,2]))    # Output: [5,6,8]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : only 1 element → outer runs once, inner runs once
# avg   → O(n^2) : inner loop runs 1+2+3+...+n = n*(n+1)/2 times = O(n^2)
# worst → O(n^2) : large array → inner always recalculates from scratch each time
# space → O(n)   : result list stores n running sums

#best time complexity    - O(n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(n)
````

````python
#Optimized Approach (Single Pass Prefix Sum)
def running_sum_optimized(nums):
    result = []
    total = 0
    for num in nums:        # single pass O(n)
        total += num        # carry forward previous sum O(1)
        result.append(total)# append current running total
    return result

print(running_sum_optimized([1,2,3,4]))  # Output: [1,3,6,10]
print(running_sum_optimized([5,1,2]))    # Output: [5,6,8]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : one pass, one addition per element → n * O(1) = O(n)
# avg   → O(n) : no branching, no nested loops, always exactly n iterations
# worst → O(n) : large n still only one pass, total carries over so no recompute
# space → O(n) : result list stores n elements (output space, unavoidable)

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

# 11. Find the Missing Number (Difficulty: 🟢 Low)

Given an array nums containing n distinct numbers in the range [0, n], return the only number in the range that is missing from the array.

Example 1:

Input: nums = [3,0,1]
Output: 2

Example 2:

Input: nums = [0,1]
Output: 2

Example 3:
Input: nums = [9,6,4,2,3,5,7,0,1]
Output: 8

````python
#Brute Force Approach
def missing_number_brute(nums):
    n = len(nums)
    for i in range(n + 1):       # outer loop: n+1 iterations
        if i not in nums:        # 'in' scans entire list → O(n) each time
            return i
    return -1

print(missing_number_brute([3,0,1]))              # Output: 2
print(missing_number_brute([0,1]))                # Output: 2
print(missing_number_brute([9,6,4,2,3,5,7,0,1])) # Output: 8

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : missing number is 0 → found on first iteration of outer loop,
#                  but 'in nums' still scans all n → O(n)
# avg   → O(n^2) : missing number is ~n/2 → ~n/2 outer iterations × n inner scan
# worst → O(n^2) : missing number is n (last) → n+1 outer × n inner = O(n^2)
# space → O(1)   : no extra storage, just loop variable i

#best time complexity    - O(n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python

#Optimized Approach (Gauss Sum Formula)
def missing_number_optimized(nums):
    n = len(nums)
    expected = n * (n + 1) // 2  # sum of 0..n using Gauss formula → O(1)
    return expected - sum(nums)   # sum(nums) is O(n), subtraction is O(1)

print(missing_number_optimized([3,0,1]))              # Output: 2
print(missing_number_optimized([0,1]))                # Output: 2
print(missing_number_optimized([9,6,4,2,3,5,7,0,1])) # Output: 8

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : sum(nums) always iterates all n elements, no shortcut possible
# avg   → O(n) : Gauss formula is O(1), sum() is O(n) → dominated by sum()
# worst → O(n) : large n → sum() still just one pass through n elements
# space → O(1) : only two integer variables (expected, result), no structures

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

# 12. Move All Zeros to the End - (Difficulty: 🟢 Low)

Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.
You must do this in-place without making a copy of the array.

Example 1:

Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

Example 2:

Input: nums = [0]
Output: [0]

````python
#Brute Force Approach
def move_zeros_brute(nums):
    non_zeros = [x for x in nums if x != 0]  # O(n) pass, O(n) space
    zeros     = [x for x in nums if x == 0]  # O(n) pass, O(n) space
    result = non_zeros + zeros               # O(n) concatenation
    for i in range(len(nums)):               # O(n) copy back
        nums[i] = result[i]
    return nums

print(move_zeros_brute([0,1,0,3,12]))  # Output: [1,3,12,0,0]
print(move_zeros_brute([0]))           # Output: [0]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : no zeros present → non_zeros=nums, zeros=[], concat+copy = O(n)
# avg   → O(n) : always 3 full passes (non_zeros, zeros, copy) → 3n = O(n)
# worst → O(n) : all zeros → same 3 passes, just non_zeros=[], zeros=nums
# space → O(n) : non_zeros + zeros together hold all n elements

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

````python
#Optimized Approach (Two Pointer In-Place)
def move_zeros_optimized(nums):
    write = 0
    for read in range(len(nums)):   # first pass: move non-zeros forward
        if nums[read] != 0:
            nums[write] = nums[read]
            write += 1
    while write < len(nums):        # second pass: fill rest with zeros
        nums[write] = 0
        write += 1
    return nums

print(move_zeros_optimized([0,1,0,3,12]))  # Output: [1,3,12,0,0]
print(move_zeros_optimized([0]))           # Output: [0]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : no zeros → write=read throughout, second while never runs,
#                but first loop still runs n times → O(n)
# avg   → O(n) : first loop always n steps, second while runs (# of zeros) steps,
#                total = n + zeros ≤ 2n = O(n)
# worst → O(n) : all zeros → first loop n steps (no writes), second loop n steps
# space → O(1) : only 'write' pointer variable, modification is in-place

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

# 13. Binary Search on a Sorted List - (Difficulty: 🟡 Medium)

Given an array of integers nums which is sorted in ascending order, and an integer target, write a function to search target in nums. If target exists, then return its index. Otherwise, return -1.
```
Example 1:

Input: nums = [-1,0,3,5,9,12], target = 9
Output: 4
```
```
Example 2:
Input: nums = [-1,0,3,5,9,12], target = 2
Output: -1

````python
#Brute Force Approach (Linear Search)
def binary_search_brute(nums, target):
    for i in range(len(nums)):   # scan left to right
        if nums[i] == target:    # O(1) comparison
            return i             # return immediately on match
    return -1

print(binary_search_brute([-1,0,3,5,9,12], 9))  # Output: 4
print(binary_search_brute([-1,0,3,5,9,12], 2))  # Output: -1

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1)   : target is first element → found in 1 comparison
# avg   → O(n)   : target at middle on average → ~n/2 comparisons = O(n)
# worst → O(n)   : target is last element or missing → scan all n elements
# space → O(1)   : only loop index i, no extra data structures

#best time complexity    - O(1)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

````python
#Optimized Approach (Binary Search)
def binary_search_optimized(nums, target):
    left, right = 0, len(nums) - 1
    while left <= right:                       # at most log2(n) iterations
        mid = left + (right - left) // 2       # avoids integer overflow
        if nums[mid] == target:
            return mid                         # found
        elif nums[mid] < target:
            left = mid + 1                     # discard left half
        else:
            right = mid - 1                    # discard right half
    return -1

print(binary_search_optimized([-1,0,3,5,9,12], 9))  # Output: 4
print(binary_search_optimized([-1,0,3,5,9,12], 2))  # Output: -1

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1)     : target is exactly the middle element → found in 1 step
# avg   → O(log n) : each iteration halves the search space:
#                    n → n/2 → n/4 → ... → 1, takes log2(n) steps
# worst → O(log n) : target not present → keep halving until left > right,
#                    after log2(n) halvings the array is exhausted
# space → O(1)     : only left, right, mid integer variables, no recursion stack

#best time complexity    - O(1)
#average time complexity - O(log n)
#worst time complexity   - O(log n)
#space complexity        - O(1)
````

# 14. Two Sum (Return Indices) - (Difficulty: 🟢 Low)

Given an array of integers nums and an integer target, return indices of the two numbers that add up to target. Assume exactly one solution.
```
Example 1
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
```
```
Example 2
Input: nums = [3,2,4], target = 6
Output: [1,2]

````python
#Brute Force Approach
def two_sum_brute(nums, target):
    for i in range(len(nums)):            # outer loop: n iterations
        for j in range(i + 1, len(nums)): # inner loop: n-i-1 iterations
            if nums[i] + nums[j] == target:
                return [i, j]             # early exit on first match
    return []

print(two_sum_brute([2,7,11,15], 9))  # Output: [0,1]
print(two_sum_brute([3,2,4], 6))      # Output: [1,2]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1)   : answer is very first pair (i=0, j=1) → found in 1 check
# avg   → O(n^2) : answer somewhere in middle → ~n^2/4 comparisons on average
# worst → O(n^2) : answer at last pair or missing → n*(n-1)/2 comparisons
# space → O(1)   : no extra structures, only two loop index variables

#best time complexity    - O(1)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python
#Optimized Approach (HashMap)
def two_sum_optimized(nums, target):
    seen = {}                          # stores {value: index}
    for i, num in enumerate(nums):     # single pass O(n)
        complement = target - num      # what value do we need?
        if complement in seen:         # O(1) hash lookup
            return [seen[complement], i]
        seen[num] = i                  # store current num and its index
    return []

print(two_sum_optimized([2,7,11,15], 9))  # Output: [0,1]
print(two_sum_optimized([3,2,4], 6))      # Output: [1,2]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1)   : complement found on second element → 1 hash lookup to return
# avg   → O(n)   : complement found around middle → ~n/2 iterations
# worst → O(n)   : answer at very end → full n iterations through array
# space → O(n)   : seen dict grows up to n-1 entries before finding answer

#best time complexity    - O(1)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

# 15. Three Sum - ( Difficulty: 🔴 High)

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that:

i != j, i != k, and j != k
nums[i] + nums[j] + nums[k] == 0

Return the answer in any order.
The solution set must not contain duplicate triplets.
```
Example 1:
Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation:

nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0
The distinct triplets are [-1,0,1] and [-1,-1,2].
```
```
Example 2:
Input: nums = [0,1,1]
Output: []
Example 3:
Input: nums = [0,0,0]
Output: [[0,0,0]]

````python
#Brute Force Approach
def three_sum_brute(nums):
    result = set()
    n = len(nums)
    for i in range(n):                  # outer loop: n
        for j in range(i+1, n):         # middle loop: n-i
            for k in range(j+1, n):     # inner loop: n-j
                if nums[i] + nums[j] + nums[k] == 0:
                    triplet = tuple(sorted([nums[i], nums[j], nums[k]]))
                    result.add(triplet) # set handles duplicates
    return [list(t) for t in result]

print(three_sum_brute([-1,0,1,2,-1,-4]))  # Output: [[-1,-1,2],[-1,0,1]]
print(three_sum_brute([0,1,1]))           # Output: []
print(three_sum_brute([0,0,0]))           # Output: [[0,0,0]]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n^3) : three nested loops always run regardless of values
#                  n * (n-1) * (n-2) / 6 ≈ O(n^3) combinations checked
# avg   → O(n^3) : no early exit possible, all triples must be checked
# worst → O(n^3) : all unique elements → all C(n,3) = O(n^3) triples checked
# space → O(n)   : result set stores unique triplets, at most O(n) triplets

#best time complexity    - O(n^3)
#average time complexity - O(n^3)
#worst time complexity   - O(n^3)
#space complexity        - O(n)
````

````python
#Optimized Approach (Sort + Two Pointer)
def three_sum_optimized(nums):
    nums.sort()          # sort first: O(n log n)
    result = []
    for i in range(len(nums) - 2):        # fix first element
        if i > 0 and nums[i] == nums[i-1]:
            continue                       # skip duplicate first elements
        left, right = i + 1, len(nums) - 1
        while left < right:               # two pointer scan: O(n)
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                result.append([nums[i], nums[left], nums[right]])
                while left < right and nums[left] == nums[left+1]:
                    left += 1             # skip duplicate left values
                while left < right and nums[right] == nums[right-1]:
                    right -= 1            # skip duplicate right values
                left += 1
                right -= 1
            elif total < 0:
                left += 1                 # need larger sum
            else:
                right -= 1               # need smaller sum
    return result

print(three_sum_optimized([-1,0,1,2,-1,-4]))  # Output: [[-1,-1,2],[-1,0,1]]
print(three_sum_optimized([0,1,1]))           # Output: []
print(three_sum_optimized([0,0,0]))           # Output: [[0,0,0]]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n log n) : all elements same → outer loop skips duplicates fast
#                      sort dominates at O(n log n)
# avg   → O(n^2)     : sort O(n log n) + outer loop n × two pointer O(n) = O(n^2)
#                      two pointer: left+right together traverse n steps per i
# worst → O(n^2)     : no duplicates → outer runs n times, inner scans n = O(n^2)
# space → O(n)       : result stores unique triplets, sort is in-place O(log n)

#best time complexity    - O(n log n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(n)
````

# 16. Product of Array Except Self (Difficulty: 🟡 Medium)

Given an integer array nums, return an array answer such that answer[i] is equal to the product of all elements except nums[i]. Solve in O(n) time without division.
```
Example 1
Input: nums = [1,2,3,4]
Output: [24,12,8,6]
```
```
Example 2
Input: nums = [-1,1,0,-3,3]
Output: [0,0,9,0,0]

````python
#Brute Force Approach
def product_except_self_brute(nums):
    n = len(nums)
    result = []
    for i in range(n):              # outer loop: n iterations
        product = 1
        for j in range(n):          # inner loop: n iterations
            if i != j:              # skip current index
                product *= nums[j]  # multiply all others
        result.append(product)
    return result

print(product_except_self_brute([1,2,3,4]))       # Output: [24,12,8,6]
print(product_except_self_brute([-1,1,0,-3,3]))   # Output: [0,0,9,0,0]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n^2) : no early exit possible, inner loop always runs n times
#                  for each of n outer iterations → always n*n operations
# avg   → O(n^2) : every element requires computing product of n-1 others
#                  n positions × (n-1) multiplications = O(n^2)
# worst → O(n^2) : large array → full n×n double loop always executes
# space → O(n)   : result list stores n product values

#best time complexity    - O(n^2)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(n)
````

````python
#Optimized Approach (Prefix + Suffix Product)
def product_except_self_optimized(nums):
    n = len(nums)
    result = [1] * n

    # Pass 1: fill result[i] with product of all elements LEFT of i
    prefix = 1
    for i in range(n):              # left to right O(n)
        result[i] = prefix
        prefix *= nums[i]           # prefix grows: 1, n[0], n[0]*n[1], ...

    # Pass 2: multiply result[i] with product of all elements RIGHT of i
    suffix = 1
    for i in range(n - 1, -1, -1): # right to left O(n)
        result[i] *= suffix
        suffix *= nums[i]           # suffix grows from right

    return result

print(product_except_self_optimized([1,2,3,4]))      # Output: [24,12,8,6]
print(product_except_self_optimized([-1,1,0,-3,3]))  # Output: [0,0,9,0,0]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : two sequential passes of n each → 2n = O(n), no shortcuts
# avg   → O(n) : prefix pass O(n) + suffix pass O(n) = O(2n) = O(n)
# worst → O(n) : large n still only two linear passes regardless of values
# space → O(n) : result array of size n (output space, unavoidable per problem)
#                only two extra scalar variables (prefix, suffix) → O(1) extra

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

# 17. Maximum Subarray Sum (Kadane's Algorithm)(Difficulty: 🟡 Medium)

Given an integer array nums, return the maximum sum of any contiguous subarray.

```
Example 1
Input: nums = [-2,1,-3,4,-1,2,1,-5,4]
Output: 6

```
```
Example 2
Input: nums = [5,4,-1,7,8]
Output: 23

````python
#Brute Force Approach
def max_subarray_brute(nums):
    max_sum = float('-inf')
    for i in range(len(nums)):          # outer: n start positions
        current_sum = 0
        for j in range(i, len(nums)):   # inner: extend subarray right
            current_sum += nums[j]
            if current_sum > max_sum:
                max_sum = current_sum
    return max_sum

print(max_subarray_brute([-2,1,-3,4,-1,2,1,-5,4]))  # Output: 6
print(max_subarray_brute([5,4,-1,7,8]))              # Output: 23

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n^2) : no early exit, inner always extends to end for each i
#                  even single max element requires checking all pairs
# avg   → O(n^2) : total iterations = n + (n-1) + ... + 1 = n(n+1)/2 = O(n^2)
# worst → O(n^2) : all negative numbers → still checks every subarray O(n^2)
# space → O(1)   : only max_sum and current_sum scalars, no structures

#best time complexity    - O(n^2)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python
#Optimized Approach (Kadane's Algorithm)
def max_subarray_optimized(nums):
    max_sum = nums[0]       # best sum seen globally
    current_sum = nums[0]   # best sum ending at current position

    for num in nums[1:]:    # single pass from index 1 O(n)
        # either extend existing subarray or start fresh from current element
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)
    return max_sum

print(max_subarray_optimized([-2,1,-3,4,-1,2,1,-5,4]))  # Output: 6
print(max_subarray_optimized([5,4,-1,7,8]))              # Output: 23

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : all positive → current_sum always grows, one full pass needed
# avg   → O(n) : single pass, two max() calls per element = 2(n-1) ops = O(n)
# worst → O(n) : all negative → current_sum resets every step, still O(n) pass
#                Kadane's key insight: at each step only 2 choices → O(1) per step
# space → O(1) : only two scalar variables (max_sum, current_sum)

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

# 18 .Subarray Sum Equals K- (Difficulty: 🔴 High)

Given an array of integers nums and an integer k, return the total number of continuous subarrays whose sum equals k.

```
Example 1:
Input: nums = [1,1,1], k = 2
Output: 2

Explanation:

[1,1] (indices 0 to 1) → sum = 2
[1,1] (indices 1 to 2) → sum = 2
```
```
Example 2:
Input: nums = [1,2,3], k = 3
Output: 2

Explanation: [1,2] and [3]

````python
#Brute Force Approach
def subarray_sum_brute(nums, k):
    count = 0
    for i in range(len(nums)):           # outer: n start positions
        current_sum = 0
        for j in range(i, len(nums)):    # inner: extend subarray O(n)
            current_sum += nums[j]
            if current_sum == k:
                count += 1              # found valid subarray
    return count

print(subarray_sum_brute([1,1,1], 2))  # Output: 2
print(subarray_sum_brute([1,2,3], 3))  # Output: 2

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n^2) : no early exit, must check all subarrays starting at each i
#                  even if k found early, other subarrays might also sum to k
# avg   → O(n^2) : outer n × inner n iterations = n(n+1)/2 total = O(n^2)
# worst → O(n^2) : array of all 1s → many matches but still full O(n^2) scan
# space → O(1)   : only count and current_sum scalars

#best time complexity    - O(n^2)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python
#Optimized Approach (Prefix Sum + HashMap)
def subarray_sum_optimized(nums, k):
    count = 0
    prefix_sum = 0
    freq = {0: 1}   # prefix_sum 0 seen once (empty subarray base case)

    for num in nums:                    # single pass O(n)
        prefix_sum += num               # running cumulative sum
        # if (prefix_sum - k) exists in freq, those subarrays sum to k
        # because: sum(i..j) = prefix[j] - prefix[i] = k
        #          → prefix[i] = prefix[j] - k
        if prefix_sum - k in freq:
            count += freq[prefix_sum - k]  # add count of valid start points
        freq[prefix_sum] = freq.get(prefix_sum, 0) + 1
    return count

print(subarray_sum_optimized([1,1,1], 2))  # Output: 2
print(subarray_sum_optimized([1,2,3], 3))  # Output: 2

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : no subarrays sum to k → still full single pass needed
# avg   → O(n) : one pass, O(1) hash ops per element → n × O(1) = O(n)
# worst → O(n) : many valid subarrays → hash still O(1) per lookup, O(n) total
#                key insight: prefix_sum - k in freq check is O(1) hash lookup
# space → O(n) : freq dict stores up to n distinct prefix sums

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

#19 . Longest Consecutive Sequence(Difficulty: 🔴 High)

Given an unsorted array of integers nums, return the length of the longest consecutive elements sequence.
You must write an algorithm that runs in O(n) time.

```
Example 1:
Input: nums = [100,4,200,1,3,2]
Output: 4

Explanation: The longest consecutive elements sequence is [1, 2, 3, 4]. Therefore its length is 4.
```
```
Example 2:
Input: nums = [0,3,7,2,5,8,4,6,0,1]
Output: 9

````python
#Brute Force Approach
def longest_consecutive_brute(nums):
    if not nums:
        return 0
    longest = 0
    for num in nums:                    # try each number as sequence start
        current = num
        streak = 1
        while current + 1 in nums:     # O(n) scan for each next number
            current += 1
            streak += 1
        longest = max(longest, streak)
    return longest

print(longest_consecutive_brute([100,4,200,1,3,2]))       # Output: 4
print(longest_consecutive_brute([0,3,7,2,5,8,4,6,0,1]))  # Output: 9

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : all same numbers → while never enters, n outer iterations
# avg   → O(n^2) : for each of n numbers, 'in nums' scans full list O(n)
#                  total = n × O(n) = O(n^2)
# worst → O(n^2) : perfectly consecutive [1,2,...,n] → first num triggers
#                  while loop n times, each 'in nums' costs O(n) = O(n^2)
# space → O(1)   : only streak and longest scalars, no extra structures

#best time complexity    - O(n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python
#Optimized Approach (HashSet + Smart Start Detection)
def longest_consecutive_optimized(nums):
    num_set = set(nums)     # O(n) to build, O(1) lookup
    longest = 0

    for num in num_set:
        # only start counting if num is the BEGINNING of a sequence
        # i.e. num-1 does NOT exist in set → avoids reprocessing middle elements
        if num - 1 not in num_set:
            current = num
            streak = 1
            while current + 1 in num_set:  # O(1) set lookup each step
                current += 1
                streak += 1
            longest = max(longest, streak)
    return longest

print(longest_consecutive_optimized([100,4,200,1,3,2]))       # Output: 4
print(longest_consecutive_optimized([0,3,7,2,5,8,4,6,0,1]))  # Output: 9

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : no consecutive sequences → outer loop n times, while never runs
# avg   → O(n) : each number processed exactly once across ALL while loops
#                the "start of sequence" check ensures no num is visited twice
#                total while iterations across all outer loops ≤ n → O(n)
# worst → O(n) : [1,2,...,n] → outer finds start at 1, while runs n times
#                but other n-1 nums skip (num-1 exists) → total = n + n = O(n)
# space → O(n) : num_set stores all n elements

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

### 20. Find the Majority Element (> n/2 times) (Difficulty: 🟡 Medium)

Given an array nums of size n, return the majority element (appears more than ⌊n / 2⌋ times). Assume it always exists.

```
Example 1
Input: nums = [3,2,3]
Output: 3

```
```
Example 2
Input: nums = [2,2,1,1,1,2,2]
Output: 2

````python
#Brute Force Approach
def majority_element_brute(nums):
    n = len(nums)
    for num in nums:                    # outer: n iterations
        count = 0
        for other in nums:              # inner: n iterations each
            if other == num:
                count += 1
        if count > n // 2:             # majority condition
            return num
    return -1

print(majority_element_brute([3,2,3]))           # Output: 3
print(majority_element_brute([2,2,1,1,1,2,2]))  # Output: 2

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : first element is majority → inner counts n, returns = O(n)
# avg   → O(n^2) : majority element somewhere in middle →
#                  ~n/2 outer × n inner = O(n^2)
# worst → O(n^2) : majority element at last position → full n×n loops
# space → O(1)   : only count scalar, no extra data structures

#best time complexity    - O(n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity
````

````python
#Optimized Approach (Boyer-Moore Voting Algorithm)
def majority_element_optimized(nums):
    candidate = None
    count = 0

    for num in nums:              # single pass O(n)
        if count == 0:
            candidate = num       # reset candidate when count hits 0
        # if num matches candidate, votes add up
        # if not, they cancel out (one majority vote vs one minority vote)
        count += 1 if num == candidate else -1

    return candidate
    # KEY INSIGHT: majority element appears > n/2 times
    # even after all cancellations, it still survives as final candidate

print(majority_element_optimized([3,2,3]))           # Output: 3
print(majority_element_optimized([2,2,1,1,1,2,2]))  # Output: 2

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : all same element → count grows to n, one pass still needed
# avg   → O(n) : single pass, two operations per element (compare + increment)
#                → 2n ops = O(n)
# worst → O(n) : alternating non-majority and majority elements →
#                count resets multiple times but pointer moves n steps total
# space → O(1) : only two variables (candidate, count) regardless of input size

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

#21 Find the Missing Number in Range 0…n (Difficulty: 🟢 Low)

Given an array nums containing n distinct numbers in the range [0, n], return the only missing number.

```
Example 1
Input: nums = [3,0,1]
Output: 2

```
```
Example 2
Input: nums = [0,1]
Output: 2

````python
#Brute Force Approach
def missing_number_brute(nums):
    n = len(nums)
    for i in range(n + 1):         # check every number from 0 to n
        if i not in nums:          # 'in' scans full list O(n) each time
            return i
    return -1

print(missing_number_brute([3,0,1]))  # Output: 2
print(missing_number_brute([0,1]))    # Output: 2

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : missing is 0 → found on first i=0 check,
#                  but 'in nums' still scans all n → O(n)
# avg   → O(n^2) : missing at ~n/2 → n/2 outer iterations × n inner scan each
# worst → O(n^2) : missing is n (last checked) → n+1 outer × n inner = O(n^2)
# space → O(1)   : no extra storage used, just loop variable i

#best time complexity    - O(n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python

#Optimized Approach (Gauss Sum Formula)
def missing_number_optimized(nums):
    n = len(nums)
    expected = n * (n + 1) // 2   # sum of 0..n using Gauss formula O(1)
    return expected - sum(nums)    # actual sum subtracted → missing number

print(missing_number_optimized([3,0,1]))  # Output: 2
print(missing_number_optimized([0,1]))    # Output: 2

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : sum(nums) always iterates all n elements, no shortcut possible
# avg   → O(n) : Gauss formula is O(1), sum() is O(n) → dominated by sum()
# worst → O(n) : large n still only one pass through all n elements
# space → O(1) : only two integer variables (expected, result)

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

# 22.Rotate Array by K Steps (Difficulty: 🟡 Medium)

Given an integer array nums, rotate the array to the right by k steps.
```
Example 1
Input: nums = [1,2,3,4,5,6,7], k = 3
Output: [5,6,7,1,2,3,4]


````python
#Brute Force Approach
def rotate_array_brute(nums, k):
    k = k % len(nums)               # normalize k to avoid full rotations
    for _ in range(k):              # perform k individual rotations
        last = nums[-1]
        for i in range(len(nums) - 1, 0, -1):  # shift all right by 1
            nums[i] = nums[i - 1]
        nums[0] = last              # place last element at front
    return nums

print(rotate_array_brute([1,2,3,4,5,6,7], 3))  # Output: [5,6,7,1,2,3,4]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : k=1 after modulo → inner shift loop runs once for n-1 steps
# avg   → O(n*k) : outer loop k times × inner loop n-1 shifts = O(n*k)
#                  when k ≈ n/2 this becomes O(n^2/2) = O(n^2)
# worst → O(n*k) : k = n//2 (largest useful k) → n/2 × n = O(n^2)
# space → O(1)   : single temp variable 'last', shifts done in-place

#best time complexity    - O(n)
#average time complexity - O(n * k)
#worst time complexity   - O(n * k)
#space complexity        - O(1)
````

````python
#Optimized Approach (Three Reverse Algorithm)
def rotate_array_optimized(nums, k):
    n = len(nums)
    k = k % n                 # normalize k

    def reverse(left, right): # helper to reverse portion of array
        while left < right:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
            right -= 1

    reverse(0, n - 1)         # step 1: reverse entire array
    reverse(0, k - 1)         # step 2: reverse first k elements
    reverse(k, n - 1)         # step 3: reverse remaining n-k elements
    return nums

print(rotate_array_optimized([1,2,3,4,5,6,7], 3))  # Output: [5,6,7,1,2,3,4]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : three reverse calls each scan at most n/2 elements
#                reverse(0,n-1): n/2 swaps, reverse(0,k-1): k/2 swaps,
#                reverse(k,n-1): (n-k)/2 swaps → total = n/2+k/2+(n-k)/2 = n
# avg   → O(n) : always exactly n/2 + k/2 + (n-k)/2 = n total swaps = O(n)
# worst → O(n) : k=n//2 → all three reverses together still do n swaps total
# space → O(1) : in-place swaps only, no extra arrays needed

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

# 23 . Merge Intervals(Difficulty: 🟡 Medium)

Given an array of intervals intervals where intervals[i] = [starti, endi], merge all overlapping intervals.
```

Example 1
Input: intervals = [[1,3],[2,6],[8,10],[15,18]]
Output: [[1,6],[8,10],[15,18]]

````python
#Brute Force Approach
def merge_intervals_brute(intervals):
    if not intervals:
        return []
    intervals.sort(key=lambda x: x[0])  # sort by start O(n log n)
    merged = [intervals[0]]

    for i in range(1, len(intervals)):   # check each interval O(n)
        for j in range(len(merged)):     # compare with every merged interval
            if intervals[i][0] <= merged[j][1]:  # overlap condition
                merged[j][1] = max(merged[j][1], intervals[i][1])
                break
        else:
            merged.append(intervals[i])
    return merged

print(merge_intervals_brute([[1,3],[2,6],[8,10],[15,18]]))
# Output: [[1,6],[8,10],[15,18]]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n log n) : no overlaps → inner loop always finds no match,
#                      sort dominates at O(n log n)
# avg   → O(n^2)     : each new interval compared with all merged intervals
#                      in worst case merged grows to n → n × n = O(n^2)
# worst → O(n^2)     : all intervals overlap → merged stays size 1 but
#                      inner loop still runs for each new interval = O(n^2)
# space → O(n)       : merged list stores up to n intervals

#best time complexity    - O(n log n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(n)
````

````python
#Optimized Approach (Sort + Single Pass Merge)
def merge_intervals_optimized(intervals):
    if not intervals:
        return []
    intervals.sort(key=lambda x: x[0])  # sort by start time O(n log n)
    merged = [intervals[0]]

    for current in intervals[1:]:        # single pass O(n)
        last = merged[-1]                # only compare with LAST merged interval
        if current[0] <= last[1]:        # overlap: current starts before last ends
            last[1] = max(last[1], current[1])  # extend end if needed
        else:
            merged.append(current)       # no overlap → add as new interval
    return merged

print(merge_intervals_optimized([[1,3],[2,6],[8,10],[15,18]]))
# Output: [[1,6],[8,10],[15,18]]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n log n) : sort O(n log n) dominates even if no merging needed
# avg   → O(n log n) : sort O(n log n) + single pass O(n) = O(n log n)
#                      after sorting, each interval compared only with last → O(1)
# worst → O(n log n) : all intervals overlap → sort still O(n log n),
#                      single pass still O(n) → dominated by sort O(n log n)
# space → O(n)       : merged list stores result intervals (output space)

#best time complexity    - O(n log n)
#average time complexity - O(n log n)
#worst time complexity   - O(n log n)
#space complexity        - O(n)
````

# 24. Find Duplicate Number (Without Modifying Array)(Difficulty: 🔴 High)

Given an array nums containing n + 1 integers where each integer is in the range [1, n], return the repeated number. Solve without modifying the array and using only constant extra space.
```
Example 1
Input: nums = [1,3,4,2,2]
Output: 2

````python
#Brute Force Approach
def find_duplicate_brute(nums):
    for i in range(len(nums)):           # outer: n iterations
        for j in range(i + 1, len(nums)):# inner: n-i iterations
            if nums[i] == nums[j]:
                return nums[i]           # first duplicate found
    return -1

print(find_duplicate_brute([1,3,4,2,2]))  # Output: 2

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n)   : duplicate is first pair (i=0,j=1) → found in 1 inner check
# avg   → O(n^2) : duplicate somewhere in middle → ~n^2/4 comparisons average
# worst → O(n^2) : duplicate only at last pair → n*(n-1)/2 comparisons = O(n^2)
# space → O(1)   : no extra data structures, only index variables

#best time complexity    - O(n)
#average time complexity - O(n^2)
#worst time complexity   - O(n^2)
#space complexity        - O(1)
````

````python
#Optimized Approach (Floyd's Cycle Detection)
def find_duplicate_optimized(nums):
    # Treat array as a linked list: index → value → next index
    # nums[i] points to nums[nums[i]], duplicate creates a cycle

    # Phase 1: detect cycle using slow/fast pointers
    slow = nums[0]
    fast = nums[0]
    while True:
        slow = nums[slow]           # move 1 step
        fast = nums[nums[fast]]     # move 2 steps
        if slow == fast:
            break                   # cycle detected

    # Phase 2: find cycle entry point (= duplicate number)
    slow = nums[0]                  # reset slow to start
    while slow != fast:
        slow = nums[slow]           # both move 1 step now
        fast = nums[fast]
    return slow                     # meeting point = duplicate

print(find_duplicate_optimized([1,3,4,2,2]))  # Output: 2

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : cycle detected quickly but phase 2 may still take O(n) steps
#                both phases traverse at most O(n) nodes of implicit linked list
# avg   → O(n) : phase 1 slow pointer moves at most n steps before meeting fast
#                phase 2 both pointers move at most n steps to find entry = O(n)
# worst → O(n) : duplicate at far end of cycle → both phases take full O(n)
# space → O(1) : only slow and fast pointer variables, array not modified

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````

# 25 . Move All Zeros to the End (In-Place)(Difficulty: 🟢 Low)

Given an integer array nums, move all 0's to the end while maintaining the relative order of non-zero elements. Do it in-place.
```
Example 1

Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

````python
#Brute Force Approach
def move_zeros_brute(nums):
    non_zeros = [x for x in nums if x != 0]   # collect non-zeros O(n)
    zeros     = [x for x in nums if x == 0]   # collect zeros O(n)
    result = non_zeros + zeros                 # combine O(n)
    for i in range(len(nums)):                 # copy back O(n)
        nums[i] = result[i]
    return nums

print(move_zeros_brute([0,1,0,3,12]))  # Output: [1,3,12,0,0]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : no zeros → non_zeros=nums, zeros=[], still 3 full O(n) passes
# avg   → O(n) : always 3 complete passes through n elements = 3n = O(n)
# worst → O(n) : all zeros → same 3 passes, non_zeros=[], zeros=nums = O(n)
# space → O(n) : non_zeros + zeros together hold all n elements in new lists

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````

````python
#Optimized Approach (Two Pointer In-Place)
def move_zeros_optimized(nums):
    write = 0                           # position to write next non-zero
    for read in range(len(nums)):       # read scans all n elements O(n)
        if nums[read] != 0:
            nums[write] = nums[read]    # move non-zero to write position
            write += 1                  # advance write pointer
    while write < len(nums):           # fill remaining positions with 0
        nums[write] = 0
        write += 1
    return nums

print(move_zeros_optimized([0,1,0,3,12]))  # Output: [1,3,12,0,0]

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n) : no zeros → write=read always, second while never runs,
#                but first loop still runs all n steps = O(n)
# avg   → O(n) : first loop n steps + second while runs (# zeros) steps
#                total = n + zeros_count ≤ 2n = O(n)
# worst → O(n) : all zeros → first loop n steps (no writes),
#                second while n steps (fill all) → 2n total = O(n)
# space → O(1) : modification entirely in-place, only write pointer variable

#best time complexity    - O(n)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(1)
````
