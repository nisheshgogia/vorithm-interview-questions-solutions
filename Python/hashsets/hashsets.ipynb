# Hashset


# Contains Duplicate
# Difficulty: 🟢 Low
Problem Statement:


Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

**Example 1:**

Input: nums = [1,2,3,1]

Output: true

Explanation: The value 1 appears at indices 0 and 3
```
**Example 2:**
```

Input: nums = [1,2,3,4]

Output: false

Explanation: All elements are distinct
```

**Example 3:**
```
Input: nums = [1,1,1,3,3,4,3,2,4,2]

Output: true

Explanation: Multiple duplicates: 1, 3, 4, and 2 all appear more than once

```

Example 4:
```

Input: nums = []

Output: false

Explanation: Empty array has no duplicates

```

Example 5:
```

Input: nums = [1]

Output: false

Explanation: Single element cannot be a duplicate

Constraints:

1 ≤ nums.length ≤ 10^5
-10^9 ≤ nums[i] ≤ 10^9

````python
#Approach 2 (Sorting First)
def contains_duplicate_sort(nums):
    if len(nums) <= 1:
        return False

    # Sort the array
    sorted_nums = sorted(nums)

    # Check adjacent elements
    for i in range(len(sorted_nums) - 1):
        if sorted_nums[i] == sorted_nums[i + 1]:
            return True

    return False

print(contains_duplicate_sort([1,2,3,1]))                # Output: True
print(contains_duplicate_sort([1,2,3,4]))                # Output: False

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(n log n) : even if duplicates adjacent after sorting,
#                      still must perform O(n log n) sort first
# avg   → O(n log n) : sorting dominates: O(n log n)
#                      adjacent check: O(n)
#                      total = O(n log n) + O(n) = O(n log n)
# worst → O(n log n) : no duplicates → full sort + full linear scan
#                      still O(n log n) + O(n) = O(n log n)
# space → O(n)       : sorted() creates new array of size n
#                      (in-place sort would be O(log n) for stack)

# WHY SORTING HELPS:
# After sorting, duplicates will be adjacent
# Only need to check adjacent pairs, not all pairs
# Example: [1,3,2,1] → [1,1,2,3] → duplicates now adjacent

# VISUALIZATION for [1,2,3,1]:
# Sort: [1,2,3,1] → [1,1,2,3]
# Check pairs: (1,1) → EQUAL! Return True ✓

#best time complexity    - O(n log n)
#average time complexity - O(n log n)
#worst time complexity   - O(n log n)
#space complexity        - O(n)


````

````python
#Optimized Approach (HashSet - Most Efficient)
def contains_duplicate_optimized(nums):
    seen = set()

    for num in nums:
        if num in seen:
            return True  # Found duplicate
        seen.add(num)

    return False

# Alternative: One-liner using set length
def contains_duplicate_oneliner(nums):
    return len(nums) != len(set(nums))

print(contains_duplicate_optimized([1,2,3,1]))             # Output: True
print(contains_duplicate_optimized([1,2,3,4]))             # Output: False
print(contains_duplicate_oneliner([1,1,1,3,3,4,3,2,4,2]))  # Output: True

# HOW COMPLEXITIES ARE DERIVED:
# best  → O(1)   : first element repeated immediately (e.g., [1,1,...])
#                  detected on second iteration
# avg   → O(n)   : duplicate found around middle of array
#                  on average, process ~n/2 elements
#                  each "in" check and add() is O(1) average for hash set
#                  total ≈ n/2 × O(1) = O(n)
# worst → O(n)   : no duplicates or duplicate at very end
#                  must process all n elements
#                  each element: O(1) check + O(1) add = O(1)
#                  total = n × O(1) = O(n)
# space → O(n)   : seen set stores up to n unique elements (all distinct)

# WHY HASHSET IS OPTIMAL:
# - O(1) average lookup time (vs O(n) for list)
# - O(1) average insertion time
# - Early termination when duplicate found

# VISUALIZATION for [1,2,3,1]:
# Iteration 1: num=1, not in seen, seen={1}
# Iteration 2: num=2, not in seen, seen={1,2}
# Iteration 3: num=3, not in seen, seen={1,2,3}
# Iteration 4: num=1, IN SEEN! Return True ✓

# ONE-LINER EXPLANATION:
# If array has duplicates, set removes them
# So len(set) < len(array) means duplicates exist
# Example: [1,2,3,1] → set={1,2,3} → len(4) != len(3) → True

#best time complexity    - O(1)
#average time complexity - O(n)
#worst time complexity   - O(n)
#space complexity        - O(n)
````
